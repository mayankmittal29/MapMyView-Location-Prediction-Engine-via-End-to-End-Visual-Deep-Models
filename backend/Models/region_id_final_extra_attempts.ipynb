```python
import os
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import gc
import random
from torch.optim.lr_scheduler import CosineAnnealingWarmRestarts

# Set seeds for reproducibility
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed()

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Define paths
TRAIN_IMG_DIR = '/kaggle/input/iiith-campus-images-dataset/images_train/images_train/images_train'
VALID_IMG_DIR = '/kaggle/input/iiith-campus-images-dataset/images_val/images_val'
TRAIN_LABELS = '/kaggle/input/iiith-campus-images-dataset/labels_train_updated.csv'
VALID_LABELS = '/kaggle/input/iiith-campus-images-dataset/labels_val_updated.csv'

# Define augmentations - much more comprehensive for training data
train_transform = transforms.Compose([
    transforms.Resize((256, 256)),  # Slightly larger than needed for random cropping
    transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.1),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),
    transforms.RandomAutocontrast(p=0.2),
    transforms.RandomEqualize(p=0.1),
    transforms.RandomAdjustSharpness(sharpness_factor=2, p=0.3),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    transforms.RandomErasing(p=0.2)
])

# Simpler transforms for validation
valid_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Improved Campus Dataset class
class CampusDataset(Dataset):
    def __init__(self, img_dir, labels_file, transform=None, is_validation=False):
        self.img_dir = img_dir
        self.transform = transform
        self.is_validation = is_validation
        self.labels_df = pd.read_csv(labels_file)
        self.valid_indices = []
        
        # Check which images can be loaded successfully
        print("Verifying dataset integrity...")
        for idx in tqdm(range(len(self.labels_df))):
            try:
                img_path = os.path.join(img_dir, self.labels_df.iloc[idx]['filename'])
                if os.path.exists(img_path):
                    with Image.open(img_path) as img:
                        if img.mode != 'RGB':
                            img = img.convert('RGB')
                        self.valid_indices.append(idx)
            except Exception as e:
                print(f"Skipping image at index {idx}: {e}")
                
        print(f"Found {len(self.valid_indices)} valid images out of {len(self.labels_df)}")

    def __len__(self):
        return len(self.valid_indices)

    def __getitem__(self, idx):
        true_idx = self.valid_indices[idx]
        img_name = os.path.join(self.img_dir, self.labels_df.iloc[true_idx]['filename'])
        
        # Use context manager for better resource handling
        with Image.open(img_name) as image:
            image = image.convert('RGB')
            
            if self.transform:
                image_tensor = self.transform(image)
                
        region_id = self.labels_df.iloc[true_idx]['Region_ID']
        
        # For validation set, also return the image filename
        if self.is_validation:
            return image_tensor, region_id - 1, self.labels_df.iloc[true_idx]['filename']
        else:
            return image_tensor, region_id - 1

# Load the datasets with appropriate transforms
try:
    train_dataset = CampusDataset(TRAIN_IMG_DIR, TRAIN_LABELS, transform=train_transform)
    valid_dataset = CampusDataset(VALID_IMG_DIR, VALID_LABELS, transform=valid_transform, is_validation=True)

    print(f"Loaded {len(train_dataset)} training samples and {len(valid_dataset)} validation samples")

    # Create data loaders with optimized parameters
    train_loader = DataLoader(
        train_dataset,
        batch_size=24,  # Smaller batch size for ConvNext (it's larger than ResNet)
        shuffle=True,
        num_workers=4,
        pin_memory=True,
        persistent_workers=True if torch.__version__ >= '1.7.0' else False,
        drop_last=True  # Drop incomplete batches for consistent batch norm stats
    )
    
    valid_loader = DataLoader(
        valid_dataset,
        batch_size=32,  # Also reduced for ConvNext
        shuffle=False,
        num_workers=4,
        pin_memory=True,
        persistent_workers=True if torch.__version__ >= '1.7.0' else False
    )

except Exception as e:
    print(f"Error loading datasets: {e}")
    print("Please check the paths and structure of your data files")

# Define a powerful model using ConvNext as base with advanced dropout and normalization techniques
class ConvNextClassifier(nn.Module):
    def __init__(self, num_classes=15):
        super(ConvNextClassifier, self).__init__()
        
        # Use ConvNeXt-Small as base model
        self.convnext = models.convnext_small(pretrained=True)
        
        # Freeze early layers to prevent overfitting
        # Freeze first 6 blocks (about 1/3 of the network)
        for name, param in self.convnext.named_parameters():
            if 'features.0.' in name or 'features.1.' in name or 'features.2.' in name:
                param.requires_grad = False
        
        # The issue happens because ConvNext models return a 4D tensor
        # We need to handle this differently than in the previous implementation
        
        # Replace the classifier and properly handle the 4D tensor
        # First, get the in_features
        if hasattr(self.convnext, 'head'):
            # For newer torchvision versions
            in_features = self.convnext.head.in_features
            # Remove the original head
            self.convnext.head = nn.Identity()
        else:
            # For older torchvision versions 
            in_features = self.convnext.classifier[2].in_features if hasattr(self.convnext.classifier, '__getitem__') else self.convnext.classifier.in_features
            # Remove the original classifier
            self.convnext.classifier = nn.Identity()
        
        # Add our own classifier that handles the 4D tensor properly
        self.classifier = nn.Sequential(
            # First adapt the 4D tensor (batch, channels, height, width) to 2D
            nn.AdaptiveAvgPool2d(1),       # -> (batch, channels, 1, 1)
            nn.Flatten(1),                 # -> (batch, channels)
            nn.LayerNorm(in_features),     # Now LayerNorm will work properly
            nn.Dropout(p=0.3),
            nn.Linear(in_features, 512),
            nn.LayerNorm(512),
            nn.GELU(),
            nn.Dropout(p=0.4),
            nn.Linear(512, 256),
            nn.LayerNorm(256),
            nn.GELU(),
            nn.Dropout(p=0.2),
            nn.Linear(256, num_classes)
        )
        
        # Initialize the weights for better convergence
        self._initialize_weights()
        
    def _initialize_weights(self):
        # Initialize only the classifier weights we added
        for m in self.classifier:
            if isinstance(m, nn.Linear):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
                    
    def forward(self, x):
        # Pass through the base model
        features = self.convnext(x)
        # Pass through our classifier
        return self.classifier(features)

# Create model instance
model = ConvNextClassifier(num_classes=15)
model = model.to(device)

# Model complexity
def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f"Model has {count_parameters(model):,} trainable parameters")

# Use CrossEntropyLoss with stronger label smoothing
criterion = nn.CrossEntropyLoss(label_smoothing=0.15)  # Increased from 0.1 for better regularization

# Use AdamW optimizer with weight decay and better learning rate
optimizer = optim.AdamW(model.parameters(), lr=1e-4, weight_decay=2e-4, amsgrad=True)

# Advanced learning rate scheduler - Cosine annealing with warm restarts
# This helps escape local minima better than manual plateaus
scheduler = CosineAnnealingWarmRestarts(
    optimizer,
    T_0=5,             # Restart every 5 epochs
    T_mult=2,          # Double the restart interval after each restart
    eta_min=1e-6       # Minimum learning rate
)

# Mixed precision training for better efficiency
scaler = torch.cuda.amp.GradScaler() if torch.cuda.is_available() else None

# Improved training function with various optimizations
def train_model(model, train_loader, valid_loader, criterion, optimizer, scheduler, num_epochs=30):
    history = {
        'train_loss': [],
        'valid_loss': [],
        'valid_acc': [],
        'valid_top3_acc': [],
        'learning_rates': []
    }

    best_acc = 0.0
    best_epoch = 0
    patience = 10  # Increased patience with better scheduler
    no_improve = 0
    validation_filenames = []
    validation_predictions = []
    
    # For AMP (Automatic Mixed Precision)
    use_amp = torch.cuda.is_available()
    
    for epoch in range(num_epochs):
        # Training phase
        model.train()
        running_loss = 0.0
        total_samples = 0
        
        progress_bar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{num_epochs} [Train]')
        for batch in progress_bar:
            inputs, labels = batch
            
            # Skip empty batches
            if inputs.size(0) == 0:
                continue

            inputs, labels = inputs.to(device), labels.to(device)
            
            # Clear gradients
            optimizer.zero_grad(set_to_none=True)
            
            if use_amp:
                # Automatic mixed precision training
                with torch.cuda.amp.autocast():
                    outputs = model(inputs)
                    loss = criterion(outputs, labels)
                
                # Scale gradients and optimize
                scaler.scale(loss).backward()
                scaler.unscale_(optimizer)
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
                scaler.step(optimizer)
                scaler.update()
            else:
                # Regular training path
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                loss.backward()
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
                optimizer.step()
            
            running_loss += loss.item() * inputs.size(0)
            total_samples += inputs.size(0)
            current_lr = optimizer.param_groups[0]['lr']
            progress_bar.set_postfix({'loss': loss.item(), 'lr': current_lr})
        
        # Step the scheduler after each epoch
        scheduler.step()
        
        epoch_loss = running_loss / total_samples if total_samples > 0 else 0
        history['train_loss'].append(epoch_loss)
        history['learning_rates'].append(optimizer.param_groups[0]['lr'])

        # Clean up to free memory
        torch.cuda.empty_cache()
        gc.collect()

        # Validation phase
        model.eval()
        running_loss = 0.0
        all_preds = []
        all_labels = []
        all_filenames = []
        total_val_samples = 0
        correct_top1 = 0
        correct_top3 = 0

        with torch.no_grad():
            progress_bar = tqdm(valid_loader, desc=f'Epoch {epoch+1}/{num_epochs} [Valid]')
            for batch in progress_bar:
                inputs, labels, filenames = batch

                # Skip empty batches
                if inputs.size(0) == 0:
                    continue

                inputs, labels = inputs.to(device), labels.to(device)
                
                if use_amp:
                    with torch.cuda.amp.autocast():
                        outputs = model(inputs)
                        loss = criterion(outputs, labels)
                else:
                    outputs = model(inputs)
                    loss = criterion(outputs, labels)

                running_loss += loss.item() * inputs.size(0)
                total_val_samples += inputs.size(0)

                # Get top-1 predictions
                _, preds = torch.max(outputs, 1)
                all_preds.extend(preds.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())
                all_filenames.extend(filenames)

                # Calculate top-1 accuracy
                correct_top1 += (preds == labels).sum().item()

                # Top-3 accuracy
                _, top3_indices = outputs.topk(3, dim=1)
                batch_correct = 0
                for i in range(labels.size(0)):
                    if labels[i] in top3_indices[i]:
                        batch_correct += 1

                correct_top3 += batch_correct

        # Calculate validation metrics
        valid_loss = running_loss / total_val_samples if total_val_samples > 0 else 0
        valid_acc = correct_top1 / total_val_samples if total_val_samples > 0 else 0
        valid_top3_acc = correct_top3 / total_val_samples if total_val_samples > 0 else 0

        history['valid_loss'].append(valid_loss)
        history['valid_acc'].append(valid_acc)
        history['valid_top3_acc'].append(valid_top3_acc)

        print(f'Epoch {epoch+1}/{num_epochs} | '
              f'Train Loss: {epoch_loss:.4f} | '
              f'Valid Loss: {valid_loss:.4f} | '
              f'Valid Acc: {valid_acc:.4f} | '
              f'Valid Top-3 Acc: {valid_top3_acc:.4f} | '
              f'LR: {optimizer.param_groups[0]["lr"]:.6f}')

        # Save if it's the best model by top-1 accuracy
        if valid_acc > best_acc:
            best_acc = valid_acc
            best_epoch = epoch
            no_improve = 0
            
            # Save predictions for best model
            validation_filenames = all_filenames
            validation_predictions = all_preds
            
            torch.save(model.state_dict(), 'best_convnext_region_classifier_3.pth')
            print(f"Saved new best model with accuracy: {best_acc:.4f}")
            
            # If we reach 98% accuracy, we can stop training
            if best_acc >= 0.985:  # Slightly higher threshold
                print(f"Reached target accuracy of 98.5%! Stopping training.")
                break
        else:
            no_improve += 1
        
        # Early stopping
        if no_improve >= patience:
            print(f"Early stopping triggered after {epoch+1} epochs")
            break
            
        # Clean up after validation
        torch.cuda.empty_cache()
        gc.collect()

    print(f"Best validation accuracy: {best_acc:.4f} at epoch {best_epoch+1}")
    return history, all_preds, all_labels, validation_filenames, validation_predictions

# Train the model
history, final_preds, final_labels, val_filenames, val_predictions = train_model(
    model=model,
    train_loader=train_loader,
    valid_loader=valid_loader,
    criterion=criterion,
    optimizer=optimizer,
    scheduler=scheduler,
    num_epochs=40
)

# Print final validation accuracy and classification report
final_acc = accuracy_score(final_labels, final_preds)
print("\nFinal Validation Accuracy:", final_acc)
print("\nClassification Report:")
print(classification_report(final_labels, final_preds,
                          target_names=[f"Region {i+1}" for i in range(15)]))

# Plot training history
plt.figure(figsize=(15, 10))

# Plot losses
plt.subplot(2, 2, 1)
plt.plot(history['train_loss'], label='Train Loss')
plt.plot(history['valid_loss'], label='Valid Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()

# Plot accuracies
plt.subplot(2, 2, 2)
plt.plot(history['valid_acc'], label='Top-1 Accuracy')
plt.plot(history['valid_top3_acc'], label='Top-3 Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Validation Accuracies')
plt.legend()

# Plot learning rates
plt.subplot(2, 2, 3)
plt.plot(history['learning_rates'], marker='o')
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.title('Learning Rate Schedule')
plt.yscale('log')

# Plot confusion matrix
plt.subplot(2, 2, 4)
cm = confusion_matrix(final_labels, final_preds)
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = np.arange(15)
plt.xticks(tick_marks, [f"{i+1}" for i in range(15)], rotation=45)
plt.yticks(tick_marks, [f"{i+1}" for i in range(15)])
plt.xlabel('Predicted Region')
plt.ylabel('True Region')

plt.tight_layout()
plt.show()

# Load best model for predictions
model.load_state_dict(torch.load('best_convnext_region_classifier_3.pth'))
model.eval()

# Create prediction function
def predict(model, image_path, transform, device):
    """Make predictions on a single image"""
    model.eval()
    with torch.no_grad():
        with Image.open(image_path) as image:
            image = image.convert('RGB')
            img_tensor = transform(image).unsqueeze(0).to(device)
            
            if torch.cuda.is_available():
                with torch.cuda.amp.autocast():
                    outputs = model(img_tensor)
            else:
                outputs = model(img_tensor)
                
            # Get probabilities and top class
            probs = torch.softmax(outputs, dim=1)
            _, preds = torch.max(outputs, 1)
            return preds.item() + 1, probs  # Add 1 to get Region_ID from 1-15

# Load validation dataset again to get filenames and create mapping to IDs
val_df = pd.read_csv(VALID_LABELS)

# Create submission dataframe with confidence scores
submission_data = []

# Add validation set predictions (ids 0 to 368)
for i, filename in enumerate(val_filenames):
    # Find the corresponding index in val_df
    val_indices = val_df[val_df['filename'] == filename].index
    if len(val_indices) > 0:
        val_idx = val_indices[0]
        
        submission_data.append({
            'id': val_idx,
            'Region_ID': int(val_predictions[i] + 1)  # Convert 0-14 to 1-15
        })

# Fill remaining test set entries with predictions from best model or default
# You could extend this to actually predict on test images if they're available
for i in range(369, 738):
    submission_data.append({
        'id': i,
        'Region_ID': 1  # Default region
    })

# Convert to DataFrame
submission = pd.DataFrame(submission_data)

# Ensure ID column has all required values and is sorted
expected_ids = set(range(0, 738))
existing_ids = set(submission['id'])
missing_ids = expected_ids - existing_ids

# Add any missing IDs with default Region_ID=1
for missing_id in missing_ids:
    submission = pd.concat([submission, pd.DataFrame([{'id': missing_id, 'Region_ID': 1}])], ignore_index=True)

# Sort by ID and reset index
submission = submission.sort_values('id').reset_index(drop=True)

# Save to CSV
submission.to_csv('2022101094_convnext.csv', index=False)
print("Submission file '2022101094_convnext.csv' created successfully with", len(submission), "entries")
```

```python
import os
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import gc
import random
import torch.nn.functional as F
from torch.optim.lr_scheduler import OneCycleLR
import timm
from timm.data.auto_augment import rand_augment_transform
from torchvision.transforms.functional import InterpolationMode

# Set seeds for reproducibility
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed()

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Define paths
TRAIN_IMG_DIR = '/kaggle/input/iiith-campus-images-dataset/images_train/images_train/images_train'
VALID_IMG_DIR = '/kaggle/input/iiith-campus-images-dataset/images_val/images_val'
TRAIN_LABELS = '/kaggle/input/iiith-campus-images-dataset/labels_train_updated.csv'
VALID_LABELS = '/kaggle/input/iiith-campus-images-dataset/labels_val_updated.csv'

# Enhanced augmentations using RandAugment
train_transform = transforms.Compose([
    transforms.Resize((256, 256), interpolation=InterpolationMode.BICUBIC),
    transforms.RandomResizedCrop(224, scale=(0.75, 1.0), ratio=(0.75, 1.33), interpolation=InterpolationMode.BICUBIC),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.2),
    rand_augment_transform(
        config_str='rand-m9-n3-mstd0.5',
        hparams={'translate_const': 100, 'img_mean': (124, 116, 104)}
    ),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    transforms.RandomErasing(p=0.3, scale=(0.02, 0.2), ratio=(0.3, 3.3), value=0)
])

# Test-time augmentation transforms
tta_transforms = [
    transforms.Compose([
        transforms.Resize((224, 224), interpolation=InterpolationMode.BICUBIC),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ]),
    transforms.Compose([
        transforms.Resize((224, 224), interpolation=InterpolationMode.BICUBIC),
        transforms.RandomHorizontalFlip(p=1.0),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ]),
    transforms.Compose([
        transforms.Resize((256, 256), interpolation=InterpolationMode.BICUBIC),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ]),
    transforms.Compose([
        transforms.Resize((256, 256), interpolation=InterpolationMode.BICUBIC),
        transforms.RandomCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ]),
]

# Regular validation transform
valid_transform = transforms.Compose([
    transforms.Resize((224, 224), interpolation=InterpolationMode.BICUBIC),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Improved Campus Dataset class with better error handling
class CampusDataset(Dataset):
    def __init__(self, img_dir, labels_file, transform=None, is_validation=False):
        self.img_dir = img_dir
        self.transform = transform
        self.is_validation = is_validation
        self.labels_df = pd.read_csv(labels_file)
        self.valid_indices = []
        
        # Check which images can be loaded successfully
        print("Verifying dataset integrity...")
        for idx in tqdm(range(len(self.labels_df))):
            try:
                img_path = os.path.join(img_dir, self.labels_df.iloc[idx]['filename'])
                if os.path.exists(img_path):
                    with Image.open(img_path) as img:
                        if img.mode != 'RGB':
                            img = img.convert('RGB')
                        self.valid_indices.append(idx)
            except Exception as e:
                print(f"Skipping image at index {idx}: {e}")
                
        print(f"Found {len(self.valid_indices)} valid images out of {len(self.labels_df)}")
        
        # Calculate class weights for focal loss if this is training data
        if not is_validation:
            # Get class distribution
            class_counts = self.labels_df.iloc[self.valid_indices]['Region_ID'].value_counts().sort_index()
            total = len(self.valid_indices)
            self.class_weights = torch.FloatTensor([total / max(1, count) for count in class_counts])
            print("Class weights:", self.class_weights)
        else:
            self.class_weights = None

    def __len__(self):
        return len(self.valid_indices)

    def __getitem__(self, idx):
        true_idx = self.valid_indices[idx]
        img_name = os.path.join(self.img_dir, self.labels_df.iloc[true_idx]['filename'])
        
        # Use context manager for better resource handling
        with Image.open(img_name) as image:
            image = image.convert('RGB')
            
            if self.transform:
                image_tensor = self.transform(image)
                
        region_id = self.labels_df.iloc[true_idx]['Region_ID']
        
        # For validation set, also return the image filename
        if self.is_validation:
            return image_tensor, region_id - 1, self.labels_df.iloc[true_idx]['filename']
        else:
            return image_tensor, region_id - 1

# Focal Loss implementation for better handling of class imbalance
class FocalLoss(nn.Module):
    def __init__(self, alpha=None, gamma=2.0, reduction='mean'):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.reduction = reduction
        self.eps = 1e-6
        
    def forward(self, inputs, targets):
        logpt = F.log_softmax(inputs, dim=1)
        pt = torch.exp(logpt)
        logpt = logpt.gather(1, targets.unsqueeze(1))
        pt = pt.gather(1, targets.unsqueeze(1))
        
        # Apply alpha if provided
        if self.alpha is not None:
            alpha = self.alpha.gather(0, targets)
            logpt = logpt * alpha.unsqueeze(1)
        
        # Calculate focal loss
        loss = -((1 - pt) ** self.gamma) * logpt
        
        if self.reduction == 'mean':
            return loss.mean()
        elif self.reduction == 'sum':
            return loss.sum()
        else:
            return loss

# Load the datasets with appropriate transforms
try:
    train_dataset = CampusDataset(TRAIN_IMG_DIR, TRAIN_LABELS, transform=train_transform)
    valid_dataset = CampusDataset(VALID_IMG_DIR, VALID_LABELS, transform=valid_transform, is_validation=True)

    print(f"Loaded {len(train_dataset)} training samples and {len(valid_dataset)} validation samples")

    # Create data loaders with optimized parameters
    train_loader = DataLoader(
        train_dataset,
        batch_size=16,  # Smaller batch size for ConvNext Base
        shuffle=True,
        num_workers=4,
        pin_memory=True,
        persistent_workers=True if torch.__version__ >= '1.7.0' else False,
        drop_last=True  # Drop incomplete batches for consistent batch norm stats
    )
    
    valid_loader = DataLoader(
        valid_dataset,
        batch_size=24,  # Also reduced for ConvNext
        shuffle=False,
        num_workers=4,
        pin_memory=True,
        persistent_workers=True if torch.__version__ >= '1.7.0' else False
    )

except Exception as e:
    print(f"Error loading datasets: {e}")
    print("Please check the paths and structure of your data files")

# Enhanced model using ConvNext-Base with Stochastic Depth and LayerNorm

# Fix the EnhancedConvNextClassifier class
class EnhancedConvNextClassifier(nn.Module):
    def __init__(self, num_classes=15, stochastic_depth_prob=0.3):
        super(EnhancedConvNextClassifier, self).__init__()
        
        # Use timm for more advanced models with stochastic depth
        self.convnext = timm.create_model(
            'convnext_base', 
            pretrained=True, 
            drop_path_rate=stochastic_depth_prob,  # Add stochastic depth
            num_classes=0,                         # Remove classifier
        )
        
        # Get the correct feature dimension
        # ConvNext from timm already has the features flattened
        if hasattr(self.convnext, 'num_features'):
            self.feature_dim = self.convnext.num_features  # For newer timm versions
        else:
            self.feature_dim = self.convnext.head.fc.in_features  # Fallback

        # Create a more sophisticated classifier head - no need for pooling/flattening
        self.classifier = nn.Sequential(
            # Remove the AdaptiveAvgPool2d and Flatten layers here
            nn.LayerNorm(self.feature_dim),
            nn.Dropout(p=0.35),
            nn.Linear(self.feature_dim, 1024),
            nn.LayerNorm(1024),
            nn.GELU(),
            nn.Dropout(p=0.45),
            nn.Linear(1024, 512),
            nn.LayerNorm(512),
            nn.GELU(),
            nn.Dropout(p=0.25),
            nn.Linear(512, num_classes)
        )
        
        # Initialize the weights for better convergence
        self._initialize_weights()
        
    def _initialize_weights(self):
        # Initialize only the classifier weights we added
        for m in self.classifier:
            if isinstance(m, nn.Linear):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
                    
    def forward(self, x):
        # The feature extraction from timm's ConvNext already includes 
        # global pooling, so no need for adaptive pooling or flattening
        features = self.convnext(x)
        # Apply classifier
        return self.classifier(features)
# Create model instance - now with ConvNext Base
model = EnhancedConvNextClassifier(num_classes=15)
model = model.to(device)

# Model complexity
def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f"Model has {count_parameters(model):,} trainable parameters")

# Use Focal Loss with class weights for better handling of class imbalance
criterion = FocalLoss(alpha=train_dataset.class_weights.to(device), gamma=2.0)

# Use AdamW optimizer with improved weight decay setting
optimizer = optim.AdamW(model.parameters(), lr=1e-4, weight_decay=5e-4, amsgrad=True)

# Calculate total steps for OneCycleLR
total_steps = len(train_loader) * 60  # 60 epochs

# Use OneCycleLR for faster convergence
scheduler = OneCycleLR(
    optimizer,
    max_lr=1e-3,
    total_steps=total_steps,
    pct_start=0.2,      # Spend 20% of time warming up
    div_factor=25,      # Initial LR is max_lr/25
    final_div_factor=1000,  # Final LR is max_lr/25000
    anneal_strategy='cos'
)

# Mixed precision training for better efficiency
scaler = torch.cuda.amp.GradScaler() if torch.cuda.is_available() else None

# Improved training function with test-time augmentation and SWA
def train_model(model, train_loader, valid_loader, criterion, optimizer, scheduler, num_epochs=60):
    history = {
        'train_loss': [],
        'valid_loss': [],
        'valid_acc': [],
        'valid_top3_acc': [],
        'learning_rates': []
    }

    best_acc = 0.0
    best_epoch = 0
    patience = 12  # Increased patience for better exploration
    no_improve = 0
    validation_filenames = []
    validation_predictions = []
    
    # For AMP (Automatic Mixed Precision)
    use_amp = torch.cuda.is_available()
    
    # EMA model (Exponential Moving Average)
    # This helps make predictions more stable
    ema_model = None
    ema_decay = 0.999
    ema_start_epoch = 3  # Start EMA after 3 epochs
    
    for epoch in range(num_epochs):
        # Training phase
        model.train()
        running_loss = 0.0
        total_samples = 0
        
        progress_bar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{num_epochs} [Train]')
        for batch_idx, batch in enumerate(progress_bar):
            inputs, labels = batch
            
            # Skip empty batches
            if inputs.size(0) == 0:
                continue

            inputs, labels = inputs.to(device), labels.to(device)
            
            # Clear gradients
            optimizer.zero_grad(set_to_none=True)
            
            if use_amp:
                # Automatic mixed precision training
                with torch.cuda.amp.autocast():
                    outputs = model(inputs)
                    loss = criterion(outputs, labels)
                
                # Scale gradients and optimize
                scaler.scale(loss).backward()
                scaler.unscale_(optimizer)
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
                scaler.step(optimizer)
                scaler.update()
            else:
                # Regular training path
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                loss.backward()
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
                optimizer.step()
            
            # Update EMA model if after start epoch
            if epoch >= ema_start_epoch:
                if ema_model is None:
                    ema_model = copy.deepcopy(model)
                    ema_model.eval()
                else:
                    for ema_param, param in zip(ema_model.parameters(), model.parameters()):
                        ema_param.data.mul_(ema_decay).add_(param.data, alpha=1 - ema_decay)
            
            # Step the scheduler after each batch (OneCycleLR needs per-batch updates)
            scheduler.step()
            
            running_loss += loss.item() * inputs.size(0)
            total_samples += inputs.size(0)
            current_lr = optimizer.param_groups[0]['lr']
            progress_bar.set_postfix({'loss': loss.item(), 'lr': current_lr})
        
        epoch_loss = running_loss / total_samples if total_samples > 0 else 0
        history['train_loss'].append(epoch_loss)
        history['learning_rates'].append(optimizer.param_groups[0]['lr'])

        # Clean up to free memory
        torch.cuda.empty_cache()
        gc.collect()

        # Validation phase - use original model for validation during training
        model.eval()
        running_loss = 0.0
        all_preds = []
        all_labels = []
        all_filenames = []
        total_val_samples = 0
        correct_top1 = 0
        correct_top3 = 0

        with torch.no_grad():
            progress_bar = tqdm(valid_loader, desc=f'Epoch {epoch+1}/{num_epochs} [Valid]')
            for batch in progress_bar:
                inputs, labels, filenames = batch

                # Skip empty batches
                if inputs.size(0) == 0:
                    continue

                inputs, labels = inputs.to(device), labels.to(device)
                
                if use_amp:
                    with torch.cuda.amp.autocast():
                        outputs = model(inputs)
                        loss = criterion(outputs, labels)
                else:
                    outputs = model(inputs)
                    loss = criterion(outputs, labels)

                running_loss += loss.item() * inputs.size(0)
                total_val_samples += inputs.size(0)

                # Get top-1 predictions
                _, preds = torch.max(outputs, 1)
                all_preds.extend(preds.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())
                all_filenames.extend(filenames)

                # Calculate top-1 accuracy
                correct_top1 += (preds == labels).sum().item()

                # Top-3 accuracy
                _, top3_indices = outputs.topk(3, dim=1)
                batch_correct = 0
                for i in range(labels.size(0)):
                    if labels[i] in top3_indices[i]:
                        batch_correct += 1

                correct_top3 += batch_correct

        # Calculate validation metrics
        valid_loss = running_loss / total_val_samples if total_val_samples > 0 else 0
        valid_acc = correct_top1 / total_val_samples if total_val_samples > 0 else 0
        valid_top3_acc = correct_top3 / total_val_samples if total_val_samples > 0 else 0

        history['valid_loss'].append(valid_loss)
        history['valid_acc'].append(valid_acc)
        history['valid_top3_acc'].append(valid_top3_acc)

        print(f'Epoch {epoch+1}/{num_epochs} | '
              f'Train Loss: {epoch_loss:.4f} | '
              f'Valid Loss: {valid_loss:.4f} | '
              f'Valid Acc: {valid_acc:.4f} | '
              f'Valid Top-3 Acc: {valid_top3_acc:.4f} | '
              f'LR: {optimizer.param_groups[0]["lr"]:.6f}')

        # Save if it's the best model by top-1 accuracy
        if valid_acc > best_acc:
            best_acc = valid_acc
            best_epoch = epoch
            no_improve = 0
            
            # Save predictions for best model
            validation_filenames = all_filenames
            validation_predictions = all_preds
            
            # Save both regular and EMA models
            torch.save(model.state_dict(), 'best_convnext_region_classifier_4.pth')
            print(f"Saved new best model with accuracy: {best_acc:.4f}")
            
            if ema_model is not None:
                torch.save(ema_model.state_dict(), 'best_convnext_region_classifier_ema_4.pth')
                
            # If we reach 98.5% accuracy, we can stop training
            if best_acc >= 0.985:
                print(f"Reached target accuracy of 98.5%! Stopping training.")
                break
        else:
            no_improve += 1
        
        # Early stopping with increased patience
        if no_improve >= patience:
            print(f"Early stopping triggered after {epoch+1} epochs")
            break
            
        # Clean up after validation
        torch.cuda.empty_cache()
        gc.collect()

    print(f"Best validation accuracy: {best_acc:.4f} at epoch {best_epoch+1}")
    return history, all_preds, all_labels, validation_filenames, validation_predictions

# For EMA implementation
import copy

# Train the model
history, final_preds, final_labels, val_filenames, val_predictions = train_model(
    model=model,
    train_loader=train_loader,
    valid_loader=valid_loader,
    criterion=criterion,
    optimizer=optimizer,
    scheduler=scheduler,
    num_epochs=60
)

# Print final validation accuracy and classification report
final_acc = accuracy_score(final_labels, final_preds)
print("\nFinal Validation Accuracy:", final_acc)
print("\nClassification Report:")
print(classification_report(final_labels, final_preds,
                           target_names=[f"Region {i+1}" for i in range(15)]))

# Plot training history
plt.figure(figsize=(15, 10))

# Plot losses
plt.subplot(2, 2, 1)
plt.plot(history['train_loss'], label='Train Loss')
plt.plot(history['valid_loss'], label='Valid Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()

# Plot accuracies
plt.subplot(2, 2, 2)
plt.plot(history['valid_acc'], label='Top-1 Accuracy')
plt.plot(history['valid_top3_acc'], label='Top-3 Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Validation Accuracies')
plt.legend()

# Plot learning rates
plt.subplot(2, 2, 3)
plt.plot(history['learning_rates'], marker='o')
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.title('Learning Rate Schedule')
plt.yscale('log')

# Plot confusion matrix
plt.subplot(2, 2, 4)
cm = confusion_matrix(final_labels, final_preds)
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = np.arange(15)
plt.xticks(tick_marks, [f"{i+1}" for i in range(15)], rotation=45)
plt.yticks(tick_marks, [f"{i+1}" for i in range(15)])
plt.xlabel('Predicted Region')
plt.ylabel('True Region')

plt.tight_layout()
plt.show()

# Load best models for ensemble prediction
model.load_state_dict(torch.load('best_convnext_region_classifier_4.pth'))
model.eval()

# Try to load EMA model if it exists
try:
    ema_model = EnhancedConvNextClassifier(num_classes=15).to(device)
    ema_model.load_state_dict(torch.load('best_convnext_region_classifier_ema_4.pth'))
    ema_model.eval()
    use_ema = True
    print("Using EMA model for ensemble predictions")
except:
    use_ema = False
    print("EMA model not found, using only primary model")

# Create test-time augmentation prediction function
def predict_with_tta(model, image_path, tta_transforms, device, ema_model=None):
    """Make predictions with test-time augmentation"""
    models_to_use = [model]
    if ema_model is not None:
        models_to_use.append(ema_model)
        
    all_probs = []
    
    with torch.no_grad():
        with Image.open(image_path) as image:
            image = image.convert('RGB')
            
            # Apply each TTA transform and get predictions
            for transform in tta_transforms:
                img_tensor = transform(image).unsqueeze(0).to(device)
                
                for mdl in models_to_use:
                    if torch.cuda.is_available():
                        with torch.cuda.amp.autocast():
                            outputs = mdl(img_tensor)
                    else:
                        outputs = mdl(img_tensor)
                        
                    # Get probabilities
                    probs = torch.softmax(outputs, dim=1)
                    all_probs.append(probs)
    
    # Average all predictions
    avg_probs = torch.mean(torch.cat(all_probs, dim=0), dim=0)
    pred_class = torch.argmax(avg_probs).item()
    
    return pred_class + 1, avg_probs.cpu()  # Add 1 to get Region_ID from 1-15

# Load validation dataset again to get filenames and create mapping to IDs
val_df = pd.read_csv(VALID_LABELS)

# Create submission dataframe with confidence scores
submission_data = []

# Add validation set predictions (ids 0 to 368)
for i, filename in enumerate(val_filenames):
    # Find the corresponding index in val_df
    val_indices = val_df[val_df['filename'] == filename].index
    if len(val_indices) > 0:
        val_idx = val_indices[0]
        
        # Use TTA for final predictions
        img_path = os.path.join(VALID_IMG_DIR, filename)
        if os.path.exists(img_path):
            try:
                pred_class, _ = predict_with_tta(
                    model, 
                    img_path, 
                    tta_transforms, 
                    device,
                    ema_model if use_ema else None
                )
                
                submission_data.append({
                    'id': val_idx,
                    'Region_ID': int(pred_class)  # Already converted to 1-15
                })
            except Exception as e:
                print(f"Error predicting {filename}: {e}")
                # Fallback to original prediction
                submission_data.append({
                    'id': val_idx,
                    'Region_ID': int(val_predictions[i] + 1)  # Convert 0-14 to 1-15
                })
        else:
            # Use the validation predictions we already have
            submission_data.append({
                'id': val_idx,
                'Region_ID': int(val_predictions[i] + 1)  # Convert 0-14 to 1-15
            })

# Fill remaining test set entries with predictions from best model or default
# You could extend this to actually predict on test images if they're available
for i in range(369, 738):
    submission_data.append({
        'id': i,
        'Region_ID': 1  # Default region
    })

# Convert to DataFrame
submission = pd.DataFrame(submission_data)

# Ensure ID column has all required values and is sorted
expected_ids = set(range(0, 738))
existing_ids = set(submission['id'])
missing_ids = expected_ids - existing_ids

# Add any missing IDs with default Region_ID=1
for missing_id in missing_ids:
    submission = pd.concat([submission, pd.DataFrame([{'id': missing_id, 'Region_ID': 1}])], ignore_index=True)

# Sort by ID and reset index
submission = submission.sort_values('id').reset_index(drop=True)

# Save to CSV
submission.to_csv('2022101094_4.csv', index=False)
print("Submission file '2022101094_4.csv' created successfully with", len(submission), "entries")
```

```python
import os
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import gc
import random
from torch.optim.lr_scheduler import CosineAnnealingWarmRestarts

# Set seeds for reproducibility
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed()

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Define paths
TRAIN_IMG_DIR = '/kaggle/input/iiith-campus-images-dataset/images_train/images_train/images_train'
VALID_IMG_DIR = '/kaggle/input/iiith-campus-images-dataset/images_val/images_val'
TRAIN_LABELS = '/kaggle/input/iiith-campus-images-dataset/labels_train_updated.csv'
VALID_LABELS = '/kaggle/input/iiith-campus-images-dataset/labels_val_updated.csv'

# Define augmentations - much more comprehensive for training data
train_transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.RandomResizedCrop(224, scale=(0.7, 1.0)),  # More aggressive crop
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.2),  # Increased probability
    transforms.RandomRotation(20),  # Increased rotation
    transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3, hue=0.15),  # More color variation
    transforms.RandomAffine(degrees=0, translate=(0.15, 0.15), scale=(0.85, 1.15)),  # Added scaling
    transforms.RandomAutocontrast(p=0.3),
    transforms.RandomEqualize(p=0.2),
    transforms.RandomAdjustSharpness(sharpness_factor=2, p=0.4),
    transforms.RandomGrayscale(p=0.05),  # Occasional grayscale
    transforms.RandomPerspective(distortion_scale=0.2, p=0.3),  # Added perspective transform
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    transforms.RandomErasing(p=0.3, scale=(0.02, 0.2)),  # Increased probability and adjusted scale
])

# Simpler transforms for validation
valid_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Improved Campus Dataset class
class CampusDataset(Dataset):
    def __init__(self, img_dir, labels_file, transform=None, is_validation=False):
        self.img_dir = img_dir
        self.transform = transform
        self.is_validation = is_validation
        self.labels_df = pd.read_csv(labels_file)
        self.valid_indices = []
        
        # Check which images can be loaded successfully
        print("Verifying dataset integrity...")
        for idx in tqdm(range(len(self.labels_df))):
            try:
                img_path = os.path.join(img_dir, self.labels_df.iloc[idx]['filename'])
                if os.path.exists(img_path):
                    with Image.open(img_path) as img:
                        if img.mode != 'RGB':
                            img = img.convert('RGB')
                        self.valid_indices.append(idx)
            except Exception as e:
                print(f"Skipping image at index {idx}: {e}")
                
        print(f"Found {len(self.valid_indices)} valid images out of {len(self.labels_df)}")

    def __len__(self):
        return len(self.valid_indices)

    def __getitem__(self, idx):
        true_idx = self.valid_indices[idx]
        img_name = os.path.join(self.img_dir, self.labels_df.iloc[true_idx]['filename'])
        
        # Use context manager for better resource handling
        with Image.open(img_name) as image:
            image = image.convert('RGB')
            
            if self.transform:
                image_tensor = self.transform(image)
                
        region_id = self.labels_df.iloc[true_idx]['Region_ID']
        
        # For validation set, also return the image filename
        if self.is_validation:
            return image_tensor, region_id - 1, self.labels_df.iloc[true_idx]['filename']
        else:
            return image_tensor, region_id - 1

# Load the datasets with appropriate transforms
try:
    train_dataset = CampusDataset(TRAIN_IMG_DIR, TRAIN_LABELS, transform=train_transform)
    valid_dataset = CampusDataset(VALID_IMG_DIR, VALID_LABELS, transform=valid_transform, is_validation=True)

    print(f"Loaded {len(train_dataset)} training samples and {len(valid_dataset)} validation samples")

    # Create data loaders with optimized parameters
    train_loader = DataLoader(
        train_dataset,
        batch_size=24,  # Smaller batch size for ConvNext (it's larger than ResNet)
        shuffle=True,
        num_workers=4,
        pin_memory=True,
        persistent_workers=True if torch.__version__ >= '1.7.0' else False,
        drop_last=True  # Drop incomplete batches for consistent batch norm stats
    )
    
    valid_loader = DataLoader(
        valid_dataset,
        batch_size=32,  # Also reduced for ConvNext
        shuffle=False,
        num_workers=4,
        pin_memory=True,
        persistent_workers=True if torch.__version__ >= '1.7.0' else False
    )

except Exception as e:
    print(f"Error loading datasets: {e}")
    print("Please check the paths and structure of your data files")

# Define a powerful model using ConvNext as base with advanced dropout and normalization techniques
class ConvNextClassifier(nn.Module):
    def __init__(self, num_classes=15):
        super(ConvNextClassifier, self).__init__()
        
        # Use ConvNeXt-Small as base model
        self.convnext = models.convnext_small(pretrained=True)
        
        # Freeze early layers to prevent overfitting
        # Freeze first 6 blocks (about 1/3 of the network)
        for name, param in self.convnext.named_parameters():
            if 'features.0.' in name or 'features.1.' in name or 'features.2.' in name:
                param.requires_grad = False
        
        # The issue happens because ConvNext models return a 4D tensor
        # We need to handle this differently than in the previous implementation
        
        # Replace the classifier and properly handle the 4D tensor
        # First, get the in_features
        if hasattr(self.convnext, 'head'):
            # For newer torchvision versions
            in_features = self.convnext.head.in_features
            # Remove the original head
            self.convnext.head = nn.Identity()
        else:
            # For older torchvision versions 
            in_features = self.convnext.classifier[2].in_features if hasattr(self.convnext.classifier, '__getitem__') else self.convnext.classifier.in_features
            # Remove the original classifier
            self.convnext.classifier = nn.Identity()
        
        # Add our own classifier that handles the 4D tensor properly
        self.classifier = nn.Sequential(
            # First adapt the 4D tensor (batch, channels, height, width) to 2D
            nn.AdaptiveAvgPool2d(1),       # -> (batch, channels, 1, 1)
            nn.Flatten(1),                 # -> (batch, channels)
            nn.LayerNorm(in_features),     # Now LayerNorm will work properly
            nn.Dropout(p=0.3),
            nn.Linear(in_features, 512),
            nn.LayerNorm(512),
            nn.GELU(),
            nn.Dropout(p=0.4),
            nn.Linear(512, 256),
            nn.LayerNorm(256),
            nn.GELU(),
            nn.Dropout(p=0.2),
            nn.Linear(256, num_classes)
        )
        
        # Initialize the weights for better convergence
        self._initialize_weights()
        
    def _initialize_weights(self):
        # Initialize only the classifier weights we added
        for m in self.classifier:
            if isinstance(m, nn.Linear):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
                if m.bias is not None:
                    nn.init.constant_(m.bias, 0)
                    
    def forward(self, x):
        # Pass through the base model
        features = self.convnext(x)
        # Pass through our classifier
        return self.classifier(features)

# Create model instance
model = ConvNextClassifier(num_classes=15)
model = model.to(device)

# Model complexity
def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f"Model has {count_parameters(model):,} trainable parameters")

# Use CrossEntropyLoss with stronger label smoothing
criterion = nn.CrossEntropyLoss(label_smoothing=0.15)  # Increased from 0.1 for better regularization

# Use AdamW optimizer with weight decay and better learning rate
optimizer = optim.AdamW(model.parameters(), lr=1e-4, weight_decay=2e-4, amsgrad=True)

# Advanced learning rate scheduler - Cosine annealing with warm restarts
# This helps escape local minima better than manual plateaus
scheduler = CosineAnnealingWarmRestarts(
    optimizer,
    T_0=5,             # Restart every 5 epochs
    T_mult=2,          # Double the restart interval after each restart
    eta_min=1e-6       # Minimum learning rate
)

# Mixed precision training for better efficiency
scaler = torch.cuda.amp.GradScaler() if torch.cuda.is_available() else None

# Improved training function with various optimizations
def train_model(model, train_loader, valid_loader, criterion, optimizer, scheduler, num_epochs=40):
    history = {
        'train_loss': [],
        'valid_loss': [],
        'valid_acc': [],
        'valid_top3_acc': [],
        'learning_rates': []
    }

    best_acc = 0.0
    best_epoch = 0
    patience = 10  # Increased patience with better scheduler
    no_improve = 0
    validation_filenames = []
    validation_predictions = []
    
    # For AMP (Automatic Mixed Precision)
    use_amp = torch.cuda.is_available()
    
    for epoch in range(num_epochs):
        # Training phase
        model.train()
        running_loss = 0.0
        total_samples = 0
        
        progress_bar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{num_epochs} [Train]')
        for batch in progress_bar:
            inputs, labels = batch
            
            # Skip empty batches
            if inputs.size(0) == 0:
                continue

            inputs, labels = inputs.to(device), labels.to(device)
            
            # Clear gradients
            optimizer.zero_grad(set_to_none=True)
            
            if use_amp:
                # Automatic mixed precision training
                with torch.cuda.amp.autocast():
                    outputs = model(inputs)
                    loss = criterion(outputs, labels)
                
                # Scale gradients and optimize
                scaler.scale(loss).backward()
                scaler.unscale_(optimizer)
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
                scaler.step(optimizer)
                scaler.update()
            else:
                # Regular training path
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                loss.backward()
                torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
                optimizer.step()
            
            running_loss += loss.item() * inputs.size(0)
            total_samples += inputs.size(0)
            current_lr = optimizer.param_groups[0]['lr']
            progress_bar.set_postfix({'loss': loss.item(), 'lr': current_lr})
        
        # Step the scheduler after each epoch
        scheduler.step()
        
        epoch_loss = running_loss / total_samples if total_samples > 0 else 0
        history['train_loss'].append(epoch_loss)
        history['learning_rates'].append(optimizer.param_groups[0]['lr'])

        # Clean up to free memory
        torch.cuda.empty_cache()
        gc.collect()

        # Validation phase
        model.eval()
        running_loss = 0.0
        all_preds = []
        all_labels = []
        all_filenames = []
        total_val_samples = 0
        correct_top1 = 0
        correct_top3 = 0

        with torch.no_grad():
            progress_bar = tqdm(valid_loader, desc=f'Epoch {epoch+1}/{num_epochs} [Valid]')
            for batch in progress_bar:
                inputs, labels, filenames = batch

                # Skip empty batches
                if inputs.size(0) == 0:
                    continue

                inputs, labels = inputs.to(device), labels.to(device)
                
                if use_amp:
                    with torch.cuda.amp.autocast():
                        outputs = model(inputs)
                        loss = criterion(outputs, labels)
                else:
                    outputs = model(inputs)
                    loss = criterion(outputs, labels)

                running_loss += loss.item() * inputs.size(0)
                total_val_samples += inputs.size(0)

                # Get top-1 predictions
                _, preds = torch.max(outputs, 1)
                all_preds.extend(preds.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())
                all_filenames.extend(filenames)

                # Calculate top-1 accuracy
                correct_top1 += (preds == labels).sum().item()

                # Top-3 accuracy
                _, top3_indices = outputs.topk(3, dim=1)
                batch_correct = 0
                for i in range(labels.size(0)):
                    if labels[i] in top3_indices[i]:
                        batch_correct += 1

                correct_top3 += batch_correct

        # Calculate validation metrics
        valid_loss = running_loss / total_val_samples if total_val_samples > 0 else 0
        valid_acc = correct_top1 / total_val_samples if total_val_samples > 0 else 0
        valid_top3_acc = correct_top3 / total_val_samples if total_val_samples > 0 else 0

        history['valid_loss'].append(valid_loss)
        history['valid_acc'].append(valid_acc)
        history['valid_top3_acc'].append(valid_top3_acc)

        print(f'Epoch {epoch+1}/{num_epochs} | '
              f'Train Loss: {epoch_loss:.4f} | '
              f'Valid Loss: {valid_loss:.4f} | '
              f'Valid Acc: {valid_acc:.4f} | '
              f'Valid Top-3 Acc: {valid_top3_acc:.4f} | '
              f'LR: {optimizer.param_groups[0]["lr"]:.6f}')

        # Save if it's the best model by top-1 accuracy
        if valid_acc > best_acc:
            best_acc = valid_acc
            best_epoch = epoch
            no_improve = 0
            
            # Save predictions for best model
            validation_filenames = all_filenames
            validation_predictions = all_preds
            
            torch.save(model.state_dict(), 'best_convnext_region_classifier_5.pth')
            print(f"Saved new best model with accuracy: {best_acc:.4f}")
            
            # If we reach 98% accuracy, we can stop training
            if best_acc >= 0.985:  # Slightly higher threshold
                print(f"Reached target accuracy of 98.5%! Stopping training.")
                break
        else:
            no_improve += 1
        
        # Early stopping
        if no_improve >= patience:
            print(f"Early stopping triggered after {epoch+1} epochs")
            break
            
        # Clean up after validation
        torch.cuda.empty_cache()
        gc.collect()

    print(f"Best validation accuracy: {best_acc:.4f} at epoch {best_epoch+1}")
    return history, all_preds, all_labels, validation_filenames, validation_predictions

# Train the model
history, final_preds, final_labels, val_filenames, val_predictions = train_model(
    model=model,
    train_loader=train_loader,
    valid_loader=valid_loader,
    criterion=criterion,
    optimizer=optimizer,
    scheduler=scheduler,
    num_epochs=40
)

# Print final validation accuracy and classification report
final_acc = accuracy_score(final_labels, final_preds)
print("\nFinal Validation Accuracy:", final_acc)
print("\nClassification Report:")
print(classification_report(final_labels, final_preds,
                          target_names=[f"Region {i+1}" for i in range(15)]))

# Plot training history
plt.figure(figsize=(15, 10))

# Plot losses
plt.subplot(2, 2, 1)
plt.plot(history['train_loss'], label='Train Loss')
plt.plot(history['valid_loss'], label='Valid Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()

# Plot accuracies
plt.subplot(2, 2, 2)
plt.plot(history['valid_acc'], label='Top-1 Accuracy')
plt.plot(history['valid_top3_acc'], label='Top-3 Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Validation Accuracies')
plt.legend()

# Plot learning rates
plt.subplot(2, 2, 3)
plt.plot(history['learning_rates'], marker='o')
plt.xlabel('Epoch')
plt.ylabel('Learning Rate')
plt.title('Learning Rate Schedule')
plt.yscale('log')

# Plot confusion matrix
plt.subplot(2, 2, 4)
cm = confusion_matrix(final_labels, final_preds)
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = np.arange(15)
plt.xticks(tick_marks, [f"{i+1}" for i in range(15)], rotation=45)
plt.yticks(tick_marks, [f"{i+1}" for i in range(15)])
plt.xlabel('Predicted Region')
plt.ylabel('True Region')

plt.tight_layout()
plt.show()

# Load best model for predictions
model.load_state_dict(torch.load('best_convnext_region_classifier_5.pth'))
model.eval()

# Create prediction function
def predict(model, image_path, transform, device):
    """Make predictions on a single image"""
    model.eval()
    with torch.no_grad():
        with Image.open(image_path) as image:
            image = image.convert('RGB')
            img_tensor = transform(image).unsqueeze(0).to(device)
            
            if torch.cuda.is_available():
                with torch.cuda.amp.autocast():
                    outputs = model(img_tensor)
            else:
                outputs = model(img_tensor)
                
            # Get probabilities and top class
            probs = torch.softmax(outputs, dim=1)
            _, preds = torch.max(outputs, 1)
            return preds.item() + 1, probs  # Add 1 to get Region_ID from 1-15

# Load validation dataset again to get filenames and create mapping to IDs
val_df = pd.read_csv(VALID_LABELS)

# Create submission dataframe with confidence scores
submission_data = []

# Add validation set predictions (ids 0 to 368)
for i, filename in enumerate(val_filenames):
    # Find the corresponding index in val_df
    val_indices = val_df[val_df['filename'] == filename].index
    if len(val_indices) > 0:
        val_idx = val_indices[0]
        
        submission_data.append({
            'id': val_idx,
            'Region_ID': int(val_predictions[i] + 1)  # Convert 0-14 to 1-15
        })

# Fill remaining test set entries with predictions from best model or default
# You could extend this to actually predict on test images if they're available
for i in range(369, 738):
    submission_data.append({
        'id': i,
        'Region_ID': 1  # Default region
    })

# Convert to DataFrame
submission = pd.DataFrame(submission_data)

# Ensure ID column has all required values and is sorted
expected_ids = set(range(0, 738))
existing_ids = set(submission['id'])
missing_ids = expected_ids - existing_ids

# Add any missing IDs with default Region_ID=1
for missing_id in missing_ids:
    submission = pd.concat([submission, pd.DataFrame([{'id': missing_id, 'Region_ID': 1}])], ignore_index=True)

# Sort by ID and reset index
submission = submission.sort_values('id').reset_index(drop=True)

# Save to CSV
submission.to_csv('2022101094_5.csv', index=False)
print("Submission file '2022101094_5.csv' created successfully with", len(submission), "entries")
```

```python
import os
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import torch.nn.functional as F

# Import pretrained models
from torchvision.models import (
    efficientnet_b0, EfficientNet_B0_Weights,
    efficientnet_b1, EfficientNet_B1_Weights,
    efficientnet_b2, EfficientNet_B2_Weights,
    convnext_tiny, ConvNeXt_Tiny_Weights,
    vit_b_16, ViT_B_16_Weights,
    swin_t, Swin_T_Weights
)

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Define paths
TRAIN_IMG_DIR = '/kaggle/input/iiith-campus-images-dataset/images_train/images_train/images_train'
VALID_IMG_DIR = '/kaggle/input/iiith-campus-images-dataset/images_val/images_val'
TRAIN_LABELS = '/kaggle/input/iiith-campus-images-dataset/labels_train_updated.csv'
VALID_LABELS = '/kaggle/input/iiith-campus-images-dataset/labels_val_updated.csv'

# Define transformations for each model type (no augmentation, just resize and normalize)
transform_efficientnet = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

transform_convnext = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

transform_vit = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

transform_swin = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Campus Dataset class
class CampusDataset(Dataset):
    def __init__(self, img_dir, labels_file, transform=None, is_validation=False):
        self.img_dir = img_dir
        self.transform = transform
        self.is_validation = is_validation
        self.labels_df = pd.read_csv(labels_file)
        self.valid_indices = []

        # Check which images can be loaded successfully
        print("Verifying dataset integrity...")
        for idx in tqdm(range(len(self.labels_df))):
            try:
                img_path = os.path.join(img_dir, self.labels_df.iloc[idx]['filename'])
                if os.path.exists(img_path):
                    Image.open(img_path).convert('RGB')
                    self.valid_indices.append(idx)
            except Exception as e:
                print(f"Skipping image at index {idx}: {e}")

        print(f"Found {len(self.valid_indices)} valid images out of {len(self.labels_df)}")

    def __len__(self):
        return len(self.valid_indices)

    def __getitem__(self, idx):
        true_idx = self.valid_indices[idx]
        img_name = os.path.join(self.img_dir, self.labels_df.iloc[true_idx]['filename'])
        image = Image.open(img_name).convert('RGB')
        region_id = self.labels_df.iloc[true_idx]['Region_ID']

        if self.transform:
            image = self.transform(image)

        # For validation set, also return the image filename
        if self.is_validation:
            return image, region_id - 1, self.labels_df.iloc[true_idx]['filename']
        else:
            return image, region_id - 1

# Load the datasets with appropriate transforms
try:
    # For EfficientNet models
    train_dataset_effnet = CampusDataset(TRAIN_IMG_DIR, TRAIN_LABELS, transform=transform_efficientnet)
    valid_dataset_effnet = CampusDataset(VALID_IMG_DIR, VALID_LABELS, transform=transform_efficientnet, is_validation=True)
    
    # For ConvNeXt
    train_dataset_convnext = CampusDataset(TRAIN_IMG_DIR, TRAIN_LABELS, transform=transform_convnext)
    valid_dataset_convnext = CampusDataset(VALID_IMG_DIR, VALID_LABELS, transform=transform_convnext, is_validation=True)
    
    # For ViT
    train_dataset_vit = CampusDataset(TRAIN_IMG_DIR, TRAIN_LABELS, transform=transform_vit)
    valid_dataset_vit = CampusDataset(VALID_IMG_DIR, VALID_LABELS, transform=transform_vit, is_validation=True)
    
    # For Swin
    train_dataset_swin = CampusDataset(TRAIN_IMG_DIR, TRAIN_LABELS, transform=transform_swin)
    valid_dataset_swin = CampusDataset(VALID_IMG_DIR, VALID_LABELS, transform=transform_swin, is_validation=True)

    print(f"Loaded {len(train_dataset_effnet)} training samples and {len(valid_dataset_effnet)} validation samples")

    # Create data loaders
    batch_size = 16
    train_loader_effnet = DataLoader(train_dataset_effnet, batch_size=batch_size, shuffle=True, num_workers=2)
    valid_loader_effnet = DataLoader(valid_dataset_effnet, batch_size=batch_size, shuffle=False, num_workers=2)
    
    train_loader_convnext = DataLoader(train_dataset_convnext, batch_size=batch_size, shuffle=True, num_workers=2)
    valid_loader_convnext = DataLoader(valid_dataset_convnext, batch_size=batch_size, shuffle=False, num_workers=2)
    
    train_loader_vit = DataLoader(train_dataset_vit, batch_size=batch_size, shuffle=True, num_workers=2)
    valid_loader_vit = DataLoader(valid_dataset_vit, batch_size=batch_size, shuffle=False, num_workers=2)
    
    train_loader_swin = DataLoader(train_dataset_swin, batch_size=batch_size, shuffle=True, num_workers=2)
    valid_loader_swin = DataLoader(valid_dataset_swin, batch_size=batch_size, shuffle=False, num_workers=2)

except Exception as e:
    print(f"Error loading datasets: {e}")
    print("Please check the paths and structure of your data files")

# Define model classes for each architecture
class EfficientNetB0Classifier(nn.Module):
    def __init__(self, num_classes=15):
        super(EfficientNetB0Classifier, self).__init__()
        weights = EfficientNet_B0_Weights.DEFAULT
        self.model = efficientnet_b0(weights=weights)
        
        # Freeze early layers
        for name, param in list(self.model.named_parameters())[:30]:
            param.requires_grad = False
            
        # Replace classifier
        in_features = self.model.classifier[1].in_features
        self.model.classifier = nn.Sequential(
            nn.Dropout(p=0.3, inplace=True),
            nn.Linear(in_features=in_features, out_features=512),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(512, num_classes)
        )

    def forward(self, x):
        return self.model(x)

class EfficientNetB1Classifier(nn.Module):
    def __init__(self, num_classes=15):
        super(EfficientNetB1Classifier, self).__init__()
        weights = EfficientNet_B1_Weights.DEFAULT
        self.model = efficientnet_b1(weights=weights)
        
        # Freeze early layers
        for name, param in list(self.model.named_parameters())[:40]:
            param.requires_grad = False
            
        # Replace classifier
        in_features = self.model.classifier[1].in_features
        self.model.classifier = nn.Sequential(
            nn.Dropout(p=0.3, inplace=True),
            nn.Linear(in_features=in_features, out_features=512),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(512, num_classes)
        )

    def forward(self, x):
        return self.model(x)

class EfficientNetB2Classifier(nn.Module):
    def __init__(self, num_classes=15):
        super(EfficientNetB2Classifier, self).__init__()
        weights = EfficientNet_B2_Weights.DEFAULT
        self.model = efficientnet_b2(weights=weights)
        
        # Freeze early layers
        for name, param in list(self.model.named_parameters())[:50]:
            param.requires_grad = False
            
        # Replace classifier
        in_features = self.model.classifier[1].in_features
        self.model.classifier = nn.Sequential(
            nn.Dropout(p=0.3, inplace=True),
            nn.Linear(in_features=in_features, out_features=512),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(512, num_classes)
        )

    def forward(self, x):
        return self.model(x)

class ConvNextClassifier(nn.Module):
    def __init__(self, num_classes=15):
        super(ConvNextClassifier, self).__init__()
        weights = ConvNeXt_Tiny_Weights.DEFAULT
        self.convnext = convnext_tiny(weights=weights)
        
        # Freeze early layers
        for name, param in self.convnext.named_parameters():
            if 'features.0' in name or 'features.1' in name or 'features.2' in name:
                param.requires_grad = False

        # Replace the classifier
        in_features = self.convnext.classifier[2].in_features
        self.convnext.classifier = nn.Sequential(
            nn.Flatten(1),
            nn.LayerNorm(in_features, eps=1e-6),
            nn.Linear(in_features, 256),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        return self.convnext(x)

class ViTClassifier(nn.Module):
    def __init__(self, num_classes=15):
        super(ViTClassifier, self).__init__()
        weights = ViT_B_16_Weights.DEFAULT
        self.vit = vit_b_16(weights=weights)
        
        # Freeze early layers (transformer blocks)
        for name, param in self.vit.named_parameters():
            if 'encoder.layers.0' in name or 'encoder.layers.1' in name or 'encoder.layers.2' in name:
                param.requires_grad = False
                
        # Replace head
        in_features = self.vit.heads.head.in_features
        self.vit.heads.head = nn.Sequential(
            nn.Linear(in_features, 512),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(512, num_classes)
        )

    def forward(self, x):
        return self.vit(x)

class SwinClassifier(nn.Module):
    def __init__(self, num_classes=15):
        super(SwinClassifier, self).__init__()
        weights = Swin_T_Weights.DEFAULT
        self.swin = swin_t(weights=weights)
        
        # Freeze early layers
        for name, param in self.swin.named_parameters():
            if 'features.0' in name or 'features.1' in name or 'features.2' in name:
                param.requires_grad = False
                
        # Replace head
        in_features = self.swin.head.in_features
        self.swin.head = nn.Sequential(
            nn.Linear(in_features, 512),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(512, num_classes)
        )

    def forward(self, x):
        return self.swin(x)

# Ensemble model that combines all individual models
class EnsembleModel(nn.Module):
    def __init__(self, models, weights=None):
        super(EnsembleModel, self).__init__()
        self.models = nn.ModuleList(models)
        # If weights are not provided, use equal weights
        self.weights = weights if weights is not None else torch.ones(len(models)) / len(models)
        self.weights = self.weights.to(device)
        
    def forward(self, x):
        # Get predictions from each model
        outputs = [model(x) for model in self.models]
        
        # Apply weights and sum
        weighted_outputs = [output * weight for output, weight in zip(outputs, self.weights)]
        ensemble_output = sum(weighted_outputs)
        
        return ensemble_output

# Training function for individual models
def train_model(model, train_loader, valid_loader, criterion, optimizer, scheduler, model_name, num_epochs=20):
    history = {
        'train_loss': [],
        'valid_loss': [],
        'valid_acc': [],
        'valid_top3_acc': []
    }

    best_acc = 0.0
    best_epoch = 0
    patience = 7  # Early stopping patience
    no_improve = 0

    for epoch in range(num_epochs):
        # Training phase
        model.train()
        running_loss = 0.0
        total_samples = 0

        progress_bar = tqdm(train_loader, desc=f'{model_name} - Epoch {epoch+1}/{num_epochs}')
        for batch in progress_bar:
            inputs, labels = batch
            
            # Skip empty batches
            if inputs.size(0) == 0:
                continue

            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            
            # Gradient clipping for stability
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)
            total_samples += inputs.size(0)
            progress_bar.set_postfix({'loss': loss.item()})

        # Update learning rate
        scheduler.step()
        
        epoch_loss = running_loss / total_samples if total_samples > 0 else 0
        history['train_loss'].append(epoch_loss)

        # Validation phase
        # In the train_model function, modify the validation loop section:

        # Validation phase
        model.eval()
        running_loss = 0.0
        all_preds = []
        all_labels = []
        total_val_samples = 0
        correct_top1 = 0
        correct_top3 = 0
        
        with torch.no_grad():
            for batch in tqdm(valid_loader, desc=f'{model_name} - Validation'):
                # Handle both training and validation batch formats
                if len(batch) == 3:  # Validation batch with filenames
                    inputs, labels, _ = batch
                else:  # Training batch
                    inputs, labels = batch
                
                # Skip empty batches
                if inputs.size(0) == 0:
                    continue

                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)

                # Rest of the validation code remains the same...

                running_loss += loss.item() * inputs.size(0)
                total_val_samples += inputs.size(0)

                # Get top-1 predictions
                _, preds = torch.max(outputs, 1)
                all_preds.extend(preds.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())

                # Calculate top-1 and top-3 accuracy
                correct_top1 += (preds == labels).sum().item()

                # Top-3 accuracy
                _, top3_indices = outputs.topk(3, dim=1)
                batch_correct = 0
                for i in range(labels.size(0)):
                    if labels[i] in top3_indices[i]:
                        batch_correct += 1

                correct_top3 += batch_correct

        # Calculate validation metrics
        valid_loss = running_loss / total_val_samples if total_val_samples > 0 else 0
        valid_acc = correct_top1 / total_val_samples if total_val_samples > 0 else 0
        valid_top3_acc = correct_top3 / total_val_samples if total_val_samples > 0 else 0

        history['valid_loss'].append(valid_loss)
        history['valid_acc'].append(valid_acc)
        history['valid_top3_acc'].append(valid_top3_acc)

        print(f'{model_name} - Epoch {epoch+1}/{num_epochs} | '
              f'Train Loss: {epoch_loss:.4f} | '
              f'Valid Loss: {valid_loss:.4f} | '
              f'Valid Acc: {valid_acc:.4f} | '
              f'Valid Top-3 Acc: {valid_top3_acc:.4f}')

        # Save the best model
        if valid_acc > best_acc:
            best_acc = valid_acc
            best_epoch = epoch
            no_improve = 0
            torch.save(model.state_dict(), f'best_{model_name}.pth')
            print(f"Saved new best {model_name} model with accuracy: {best_acc:.4f}")
        else:
            no_improve += 1
            
        # Early stopping
        if no_improve >= patience:
            print(f"Early stopping triggered for {model_name} after {epoch+1} epochs")
            break

    print(f"Best validation accuracy for {model_name}: {best_acc:.4f} at epoch {best_epoch+1}")
    return history, best_acc

# Function to get validation predictions from a model
def get_validation_predictions(model, valid_loader):
    model.eval()
    all_outputs = []
    all_labels = []
    all_filenames = []
    
    with torch.no_grad():
        for batch in tqdm(valid_loader, desc='Getting predictions'):
            inputs, labels, filenames = batch
            
            # Skip empty batches
            if inputs.size(0) == 0:
                continue
                
            inputs = inputs.to(device)
            outputs = model(inputs)
            
            all_outputs.append(outputs.cpu())
            all_labels.extend(labels.numpy())
            all_filenames.extend(filenames)
            
    return torch.cat(all_outputs), np.array(all_labels), all_filenames

# Train and optimize each model
def train_all_models():
    # Common hyperparameters
    criterion = nn.CrossEntropyLoss(label_smoothing=0.1)
    num_epochs = 20
    
    # Initialize models
    models = {
        "efficientnet_b0": EfficientNetB0Classifier(num_classes=15).to(device),
        "efficientnet_b1": EfficientNetB1Classifier(num_classes=15).to(device),
        "efficientnet_b2": EfficientNetB2Classifier(num_classes=15).to(device),
        "convnext": ConvNextClassifier(num_classes=15).to(device),
        "vit": ViTClassifier(num_classes=15).to(device),
        "swin": SwinClassifier(num_classes=15).to(device)
    }
    
    train_loaders = {
        "efficientnet_b0": train_loader_effnet,
        "efficientnet_b1": train_loader_effnet,
        "efficientnet_b2": train_loader_effnet,
        "convnext": train_loader_convnext,
        "vit": train_loader_vit,
        "swin": train_loader_swin
    }
    
    valid_loaders = {
        "efficientnet_b0": valid_loader_effnet,
        "efficientnet_b1": valid_loader_effnet,
        "efficientnet_b2": valid_loader_effnet,
        "convnext": valid_loader_convnext,
        "vit": valid_loader_vit,
        "swin": valid_loader_swin
    }
    
    # Optimizers and schedulers
    optimizers = {}
    schedulers = {}
    histories = {}
    best_accuracies = {}
    
    # Define optimizers and schedulers for each model
    for name, model in models.items():
        optimizers[name] = optim.AdamW(model.parameters(), lr=0.0001, weight_decay=1e-5)
        schedulers[name] = optim.lr_scheduler.CosineAnnealingWarmRestarts(
            optimizers[name], T_0=5, T_mult=2
        )
    
    # Train each model
    for name, model in models.items():
        print(f"\n===== Training {name} =====")
        history, best_acc = train_model(
            model=model,
            train_loader=train_loaders[name],
            valid_loader=valid_loaders[name],
            criterion=criterion,
            optimizer=optimizers[name],
            scheduler=schedulers[name],
            model_name=name,
            num_epochs=num_epochs
        )
        histories[name] = history
        best_accuracies[name] = best_acc
    
    return models, histories, best_accuracies

# Function to train ensemble with optimized weights
def train_ensemble(models, valid_loader):
    # Load best weights for each model
    for name, model in models.items():
        model.load_state_dict(torch.load(f'best_{name}.pth'))
        model.eval()  # Set models to evaluation mode
    
    # Get validation predictions from each model
    predictions = {}
    labels = None
    filenames = None
    
    for name, model in models.items():
        print(f"Getting predictions for {name}...")
        if name in ["efficientnet_b0", "efficientnet_b1", "efficientnet_b2"]:
            preds, lbls, fnames = get_validation_predictions(model, valid_loader_effnet)
        elif name == "convnext":
            preds, lbls, fnames = get_validation_predictions(model, valid_loader_convnext)
        elif name == "vit":
            preds, lbls, fnames = get_validation_predictions(model, valid_loader_vit)
        elif name == "swin":
            preds, lbls, fnames = get_validation_predictions(model, valid_loader_swin)
        
        predictions[name] = preds
        if labels is None:
            labels = lbls
            filenames = fnames
    
    # Initialize ensemble weights (starting with equal weights)
    num_models = len(models)
    ensemble_weights = nn.Parameter(torch.ones(num_models) / num_models)
    
    # Optimize ensemble weights
    optimizer = optim.Adam([ensemble_weights], lr=0.01)
    criterion = nn.CrossEntropyLoss()
    
    # Map model names to indices
    name_to_idx = {name: i for i, name in enumerate(models.keys())}
    
    # Training loop for ensemble weights
    best_acc = 0
    best_weights = ensemble_weights.data.clone()
    
    print("Optimizing ensemble weights...")
    for epoch in range(100):  # Optimize for 100 epochs
        optimizer.zero_grad()
        
        # Apply softmax to ensure weights sum to 1
        normalized_weights = F.softmax(ensemble_weights, dim=0)
        
        # Weighted sum of predictions
        ensemble_preds = torch.zeros_like(predictions["efficientnet_b0"])
        for name, preds in predictions.items():
            ensemble_preds += preds * normalized_weights[name_to_idx[name]]
        
        # Calculate loss
        loss = criterion(ensemble_preds, torch.tensor(labels).to(device))
        loss.backward()
        optimizer.step()
        
        # Evaluate ensemble
        _, preds = torch.max(ensemble_preds, 1)
        accuracy = accuracy_score(labels, preds.cpu().numpy())
        
        if accuracy > best_acc:
            best_acc = accuracy
            best_weights = normalized_weights.data.clone()
            
        if epoch % 10 == 0:
            print(f"Epoch {epoch}, Ensemble Accuracy: {accuracy:.4f}, Loss: {loss.item():.4f}")
            print(f"Weights: {normalized_weights.data.cpu().numpy()}")
    
    print(f"Best ensemble accuracy: {best_acc:.4f}")
    print(f"Best weights: {best_weights.data.cpu().numpy()}")
    
    # Create ensemble model with best weights
    model_list = list(models.values())
    ensemble = EnsembleModel(model_list, weights=best_weights).to(device)
    
    return ensemble, best_weights, filenames

# Create final predictions using the ensemble
def create_submission(ensemble, valid_loader, val_filenames):
    ensemble.eval()
    
    # Get predictions for validation set
    all_outputs = []
    all_labels = []
    
    with torch.no_grad():
        for batch in tqdm(valid_loader, desc='Getting ensemble predictions'):
            inputs, labels, _ = batch
            
            # Skip empty batches
            if inputs.size(0) == 0:
                continue
                
            inputs = inputs.to(device)
            outputs = ensemble(inputs)
            
            _, preds = torch.max(outputs, 1)
            all_outputs.extend(preds.cpu().numpy())
            all_labels.extend(labels.numpy())
    
    # Print ensemble performance
    accuracy = accuracy_score(all_labels, all_outputs)
    print(f"Final Ensemble Validation Accuracy: {accuracy:.4f}")
    print("\nClassification Report:")
    print(classification_report(all_labels, all_outputs,
                              target_names=[f"Region {i+1}" for i in range(15)]))
    
    # Plot confusion matrix
    cm = confusion_matrix(all_labels, all_outputs)
    plt.figure(figsize=(10, 8))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Ensemble Confusion Matrix')
    plt.colorbar()
    tick_marks = np.arange(15)
    plt.xticks(tick_marks, [f"Region {i+1}" for i in range(15)], rotation=45)
    plt.yticks(tick_marks, [f"Region {i+1}" for i in range(15)])
    plt.tight_layout()
    plt.xlabel('Predicted Region')
    plt.ylabel('True Region')
    plt.show()
    
    # Load validation dataset to get filenames and create mapping to IDs
    val_df = pd.read_csv(VALID_LABELS)
    
    # Create submission dataframe
    submission_data = []
    
    # Add validation set predictions (ids 0 to 368)
    for i, filename in enumerate(val_filenames):
        # Find the corresponding index in val_df
        val_idx = val_df[val_df['filename'] == filename].index[0]
        
        submission_data.append({
            'id': val_idx,
            'Region_ID': int(all_outputs[i] + 1)  # Convert 0-14 to 1-15
        })
    
    # Fill remaining test set entries with placeholder values (ids 369 to 737)
    for i in range(369, 738):
        submission_data.append({
            'id': i,
            'Region_ID': 1  # Default region as requested
        })
    
    # Convert to DataFrame
    submission = pd.DataFrame(submission_data)
    
    # Ensure ID column has all required values and is sorted
    expected_ids = set(range(0, 738))
    existing_ids = set(submission['id'])
    missing_ids = expected_ids - existing_ids
    
    # Add any missing IDs with default Region_ID=1
    for missing_id in missing_ids:
        submission = pd.concat([submission, pd.DataFrame([{'id': missing_id, 'Region_ID': 1}])], ignore_index=True)
    
    # Sort by ID and reset index
    submission = submission.sort_values('id').reset_index(drop=True)
    
    # Save to CSV
    submission.to_csv('2022101094_ensemble.csv', index=False)
    print("Submission file '2022101094_ensemble.csv' created successfully with", len(submission), "entries")

# Main execution
if __name__ == "__main__":
    # Train all individual models
    models, histories, best_accuracies = train_all_models()
    
    # Print individual model performances
    print("\n===== Individual Model Best Accuracies =====")
    for name, acc in best_accuracies.items():
        print(f"{name}: {acc:.4f}")
    
    # Create and train ensemble model
    ensemble, best_weights, val_filenames = train_ensemble(models, valid_loader_effnet)
    
    # Create submission file
    create_submission(ensemble, valid_loader_effnet, val_filenames)
    
    # Plot training histories for all models
    plt.figure(figsize=(20, 10))
    
    # Plot training loss
    plt.subplot(2, 2, 1)
    for name, history in histories.items():
        plt.plot(history['train_loss'], label=name)
    plt.title('Training Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    # Plot validation loss
    plt.subplot(2, 2, 2)
    for name, history in histories.items():
        plt.plot(history['valid_loss'], label=name)
    plt.title('Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    
    # Plot validation accuracy
    plt.subplot(2, 2, 3)
    for name, history in histories.items():
        plt.plot(history['valid_acc'], label=name)
    plt.title('Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    
# Plot validation top-3 accuracy
    plt.subplot(2, 2, 4)
    for name, history in histories.items():
        plt.plot(history['valid_top3_acc'], label=name)
    plt.title('Validation Top-3 Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    
    plt.tight_layout()
    plt.savefig('model_comparison.png')
    plt.show()
    
    # Plot ensemble weights
    plt.figure(figsize=(10, 6))
    names = list(models.keys())
    plt.bar(names, best_weights.cpu().numpy())
    plt.title('Optimized Ensemble Model Weights')
    plt.xlabel('Model')
    plt.ylabel('Weight')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    
    # Save ensemble model
    torch.save({
        'model_state_dict': ensemble.state_dict(),
        'weights': best_weights
    }, 'best_ensemble_model.pth')
    
    print("Complete training and evaluation process finished!")
    
    # Function for inference on new images
    def predict_region(image_path, ensemble_model, transforms_dict):
        """
        Predict region for a single image using the ensemble model
        
        Args:
            image_path: Path to the image file
            ensemble_model: Trained ensemble model
            transforms_dict: Dictionary of transforms for each model type
            
        Returns:
            predicted_region: Region ID (1-15)
            confidence: Confidence score
        """
        # Load and preprocess image
        image = Image.open(image_path).convert('RGB')
        
        # Create batch of different transforms for each model
        batch_inputs = []
        for transform_key, transform in transforms_dict.items():
            tensor = transform(image).unsqueeze(0).to(device)
            batch_inputs.append(tensor)
        
        # Make prediction
        ensemble_model.eval()
        with torch.no_grad():
            # Process each input with its corresponding model
            outputs = []
            for i, model in enumerate(ensemble_model.models):
                output = model(batch_inputs[i])
                outputs.append(output)
            
            # Weight and combine outputs
            weighted_outputs = [output * weight for output, weight in zip(outputs, ensemble_model.weights)]
            ensemble_output = sum(weighted_outputs)
            
            # Get prediction and confidence
            probabilities = F.softmax(ensemble_output, dim=1)
            confidence, prediction = torch.max(probabilities, dim=1)
            
        return prediction.item() + 1, confidence.item()  # Adding 1 to convert to 1-15 scale
    
    # Example usage of prediction function (for testing)
    def test_ensemble_prediction():
        # Dictionary of transforms
        transforms_dict = {
            "efficientnet": transform_efficientnet,
            "convnext": transform_convnext,
            "vit": transform_vit,
            "swin": transform_swin
        }
        
        # Load a test image from validation set
        test_img_path = os.path.join(VALID_IMG_DIR, val_df.iloc[0]['filename'])
        true_region = val_df.iloc[0]['Region_ID']
        
        # Predict
        pred_region, confidence = predict_region(test_img_path, ensemble, transforms_dict)
        
        print(f"Example prediction - Image: {val_df.iloc[0]['filename']}")
        print(f"True Region: {true_region}")
        print(f"Predicted Region: {pred_region}")
        print(f"Confidence: {confidence:.4f}")
        
    # Uncomment to test prediction on a sample image
    # test_ensemble_prediction()
    
    # To get predictions for a dataset of images
    def predict_dataset(image_dir, ensemble_model, transforms_dict):
        """
        Generate predictions for all images in a directory
        
        Args:
            image_dir: Directory containing images
            ensemble_model: Trained ensemble model
            transforms_dict: Dictionary of transforms for each model type
            
        Returns:
            predictions: Dictionary of {filename: predicted_region}
        """
        predictions = {}
        files = os.listdir(image_dir)
        image_files = [f for f in files if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        
        for img_file in tqdm(image_files, desc="Predicting regions"):
            img_path = os.path.join(image_dir, img_file)
            try:
                pred_region, _ = predict_region(img_path, ensemble_model, transforms_dict)
                predictions[img_file] = pred_region
            except Exception as e:
                print(f"Error processing {img_file}: {e}")
                predictions[img_file] = 1  # Default to region 1 on error
                
        return predictions
```

```python
# Function to train ensemble with optimized weights - Further fixed version
def train_ensemble(models, valid_loader):
    # Load best weights for each model
    for name, model in models.items():
        model.load_state_dict(torch.load(f'best_{name}.pth'))
        model.eval()  # Set models to evaluation mode
    
    # Get validation predictions from each model
    predictions = {}
    labels = None
    filenames = None
    
    for name, model in models.items():
        print(f"Getting predictions for {name}...")
        if name in ["efficientnet_b0", "efficientnet_b1", "efficientnet_b2"]:
            preds, lbls, fnames = get_validation_predictions(model, valid_loader_effnet)
        elif name == "convnext":
            preds, lbls, fnames = get_validation_predictions(model, valid_loader_convnext)
        elif name == "vit":
            preds, lbls, fnames = get_validation_predictions(model, valid_loader_vit)
        elif name == "swin":
            preds, lbls, fnames = get_validation_predictions(model, valid_loader_swin)
        
        # Move predictions to device - this is the key fix!
        predictions[name] = preds.to(device)
        
        if labels is None:
            labels = lbls
            filenames = fnames
    
    # Convert labels to a tensor and move to the same device as predictions
    labels_tensor = torch.tensor(labels, dtype=torch.long).to(device)
    
    # Initialize ensemble weights (starting with equal weights)
    num_models = len(models)
    ensemble_weights = nn.Parameter(torch.ones(num_models, device=device) / num_models)
    
    # Optimize ensemble weights
    optimizer = optim.Adam([ensemble_weights], lr=0.01)
    criterion = nn.CrossEntropyLoss()
    
    # Map model names to indices
    name_to_idx = {name: i for i, name in enumerate(models.keys())}
    
    # Training loop for ensemble weights
    best_acc = 0
    best_weights = ensemble_weights.data.clone()
    
    print("Optimizing ensemble weights...")
    for epoch in range(100):  # Optimize for 100 epochs
        optimizer.zero_grad()
        
        # Apply softmax to ensure weights sum to 1
        normalized_weights = F.softmax(ensemble_weights, dim=0)
        
        # Weighted sum of predictions - ensure all tensors are on the same device
        ensemble_preds = torch.zeros_like(predictions["efficientnet_b0"])
        for name, preds in predictions.items():
            idx = name_to_idx[name]
            weight = normalized_weights[idx]
            ensemble_preds += preds * weight
        
        # Calculate loss
        loss = criterion(ensemble_preds, labels_tensor)
        loss.backward()
        optimizer.step()
        
        # Evaluate ensemble
        _, preds = torch.max(ensemble_preds, 1)
        accuracy = accuracy_score(labels, preds.cpu().numpy())
        
        if accuracy > best_acc:
            best_acc = accuracy
            best_weights = normalized_weights.data.clone()
            
        if epoch % 10 == 0:
            print(f"Epoch {epoch}, Ensemble Accuracy: {accuracy:.4f}, Loss: {loss.item():.4f}")
            print(f"Weights: {normalized_weights.data.cpu().numpy()}")
    
    print(f"Best ensemble accuracy: {best_acc:.4f}")
    print(f"Best weights: {best_weights.data.cpu().numpy()}")
    
    # Create ensemble model with best weights
    model_list = list(models.values())
    ensemble = EnsembleModel(model_list, weights=best_weights).to(device)
    
    return ensemble, best_weights, filenames

# Create submission file
def create_submission(ensemble, valid_loader, val_filenames):
    ensemble.eval()
    
    # Get predictions for validation set
    all_outputs = []
    all_labels = []
    
    with torch.no_grad():
        for batch in tqdm(valid_loader, desc='Getting ensemble predictions'):
            inputs, labels, _ = batch
            
            # Skip empty batches
            if inputs.size(0) == 0:
                continue
                
            inputs = inputs.to(device)
            outputs = ensemble(inputs)
            
            _, preds = torch.max(outputs, 1)
            all_outputs.extend(preds.cpu().numpy())
            all_labels.extend(labels.numpy())
    
    # Print ensemble performance
    accuracy = accuracy_score(all_labels, all_outputs)
    print(f"Final Ensemble Validation Accuracy: {accuracy:.4f}")
    print("\nClassification Report:")
    print(classification_report(all_labels, all_outputs,
                              target_names=[f"Region {i+1}" for i in range(15)]))
    
    # Plot confusion matrix
    cm = confusion_matrix(all_labels, all_outputs)
    plt.figure(figsize=(10, 8))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title('Ensemble Confusion Matrix')
    plt.colorbar()
    tick_marks = np.arange(15)
    plt.xticks(tick_marks, [f"Region {i+1}" for i in range(15)], rotation=45)
    plt.yticks(tick_marks, [f"Region {i+1}" for i in range(15)])
    plt.tight_layout()
    plt.xlabel('Predicted Region')
    plt.ylabel('True Region')
    plt.show()
    
    # Load validation dataset to get filenames and create mapping to IDs
    val_df = pd.read_csv(VALID_LABELS)
    
    # Create submission dataframe
    submission_data = []
    
    # Add validation set predictions (ids 0 to 368)
    for i, filename in enumerate(val_filenames):
        # Find the corresponding index in val_df
        idx_matches = val_df[val_df['filename'] == filename].index
        if len(idx_matches) > 0:
            val_idx = idx_matches[0]
            
            submission_data.append({
                'id': val_idx,
                'Region_ID': int(all_outputs[i] + 1)  # Convert 0-14 to 1-15
            })
        else:
            print(f"Warning: Could not find {filename} in validation labels dataframe")
    
    # Fill remaining test set entries with placeholder values (ids 369 to 737)
    for i in range(369, 738):
        submission_data.append({
            'id': i,
            'Region_ID': 1  # Default region as requested
        })
    
    # Convert to DataFrame
    submission = pd.DataFrame(submission_data)
    
    # Ensure ID column has all required values and is sorted
    expected_ids = set(range(0, 738))
    existing_ids = set(submission['id'])
    missing_ids = expected_ids - existing_ids
    
    # Add any missing IDs with default Region_ID=1
    for missing_id in missing_ids:
        submission = pd.concat([submission, pd.DataFrame([{'id': missing_id, 'Region_ID': 1}])], ignore_index=True)
    
    # Sort by ID and reset index
    submission = submission.sort_values('id').reset_index(drop=True)
    
    # Save to CSV
    submission.to_csv('2022101094_ensemble.csv', index=False)
    print("Submission file '2022101094_ensemble.csv' created successfully with", len(submission), "entries")

# Main execution code
# Run this after your models have been trained

# Train ensemble model with the fixed function
try:
    ensemble, best_weights, val_filenames = train_ensemble(models, valid_loader_effnet)
    
    # Create submission file
    create_submission(ensemble, valid_loader_effnet, val_filenames)
    
    # Plot ensemble weights
    plt.figure(figsize=(10, 6))
    names = list(models.keys())
    plt.bar(names, best_weights.cpu().numpy())
    plt.title('Optimized Ensemble Model Weights')
    plt.xlabel('Model')
    plt.ylabel('Weight')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    
    # Save ensemble model
    torch.save({
        'model_state_dict': ensemble.state_dict(),
        'weights': best_weights
    }, 'best_ensemble_model.pth')
    
    print("Complete training and evaluation process finished!")
    
except Exception as e:
    print(f"Error in ensemble training: {e}")
    import traceback
    traceback.print_exc()
```

```python
def calculate_accuracy(model, data_loader, device):
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in data_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            predicted = torch.argmax(outputs, 1)
            correct += (predicted == torch.argmax(labels,1)).sum().item()
            total += labels.size(0)

    accuracy = 100 * correct / total
    print(f"Accuracy: {accuracy:.2f}%")
    return accuracy
```

```python
from efficientnet_pytorch import EfficientNet

class GeoEfficientNet_GroupOnly(nn.Module):
    def __init__(self, model_name='efficientnet-b0', num_groups=15,dropout_rate=0.3):
        super(GeoEfficientNet_GroupOnly, self).__init__()
        self.base_model = EfficientNet.from_pretrained(model_name)
        in_features = self.base_model._fc.in_features
        self.base_model._dropout = nn.Dropout(p=dropout_rate)
        self.base_model._fc = nn.Linear(in_features, num_groups)

    def forward(self, x):
        return self.base_model(x)

    def train_model(self, train_loader, val_loader, criterion, optimizer, device, epochs=10):
        self.to(device)
        best_acc = 0
        best_state_dict = None
    
        scheduler = torch.optim.lr_scheduler.CosineAnnealingWarmRestarts(optimizer, T_0=20)
    
        for epoch in range(epochs):
            self.train()
            train_loss = 0
            for images, targets in tqdm(train_loader):
                images, targets = images.to(device), targets.to(device)
                optimizer.zero_grad()
                outputs = self(images)
                loss = criterion(outputs, targets)
                loss.backward()
                optimizer.step()
                train_loss += loss.item()
            
            scheduler.step()  #  Cosine LR step here
            avg_train_loss = train_loss / len(train_loader)
    
            val_loss, accuracy = self.evaluate(val_loader, criterion, device, print_loss=False)
            print(f"Epoch {epoch+1}, Train Loss: {avg_train_loss:.4f}, Val Loss: {val_loss:.4f}")
    
            if accuracy > best_acc:
                print("Validation loss improved. Saving model.")
                best_acc = accuracy
                best_state_dict = copy.deepcopy(self.state_dict())
    
        return best_state_dict

    def evaluate(self, data_loader, criterion, device, print_loss=True):
        self.eval()
        total_loss = 0
        with torch.no_grad():
            for images, targets in data_loader:
                images, targets = images.to(device), targets.to(device)
                outputs = self(images)
                loss = criterion(outputs, targets)
                total_loss += loss.item()
        avg_loss = total_loss / len(data_loader)
        
        if print_loss:
            print(f"Cross-Entropy Loss: {avg_loss:.4f}")
            
        return avg_loss , calculate_accuracy(self,val_loader,device)
```

```python
import os
import torch
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from torchvision import transforms
from transformers import (
    ViTForImageClassification, 
    ConvNextForImageClassification,
    SwinForImageClassification
)
import timm
import torch.nn as nn
import argparse
import json

# Define device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Define number of classes
NUM_CLASSES = 15

# Validation transforms dictionary for different models
val_transforms = {
    "default": transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ]),
    "vit": transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
    ]),
    "convnext": transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ]),
    "swin": transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
}

# Model definitions (same as in the training script)
class ConvNextClassifier(nn.Module):
    def __init__(self, num_classes=NUM_CLASSES):
        super(ConvNextClassifier, self).__init__()
        self.model = ConvNextForImageClassification.from_pretrained(
            "facebook/convnext-tiny-224", 
            num_labels=num_classes,
            ignore_mismatched_sizes=True
        )
        
    def forward(self, x):
        outputs = self.model(x)
        return outputs.logits

class ViTClassifier(nn.Module):
    def __init__(self, num_classes=NUM_CLASSES):
        super(ViTClassifier, self).__init__()
        self.model = ViTForImageClassification.from_pretrained(
            "google/vit-base-patch16-224", 
            num_labels=num_classes,
            ignore_mismatched_sizes=True
        )
        
    def forward(self, x):
        outputs = self.model(x)
        return outputs.logits

class SwinClassifier(nn.Module):
    def __init__(self, num_classes=NUM_CLASSES):
        super(SwinClassifier, self).__init__()
        self.model = SwinForImageClassification.from_pretrained(
            "microsoft/swin-tiny-patch4-window7-224", 
            num_labels=num_classes,
            ignore_mismatched_sizes=True
        )
        
    def forward(self, x):
        outputs = self.model(x)
        return outputs.logits

class EfficientNetClassifier(nn.Module):
    def __init__(self, num_classes=NUM_CLASSES):
        super(EfficientNetClassifier, self).__init__()
        self.model = timm.create_model('tf_efficientnetv2_b0', pretrained=True, num_classes=0)
        self.dropout = nn.Dropout(0.3)
        self.fc = nn.Linear(1280, num_classes)
        
    def forward(self, x):
        features = self.model(x)
        x = self.dropout(features)
        x = self.fc(x)
        return x

# Model factory function
def get_model(model_name, num_classes=NUM_CLASSES):
    if model_name == "convnext":
        return ConvNextClassifier(num_classes=num_classes)
    elif model_name == "vit":
        return ViTClassifier(num_classes=num_classes)
    elif model_name == "swin":
        return SwinClassifier(num_classes=num_classes)
    elif model_name == "efficientnet":
        return EfficientNetClassifier(num_classes=num_classes)
    else:
        raise ValueError(f"Unknown model: {model_name}")

# Function to load a single model
def load_model(model_name, weights_path):
    model = get_model(model_name)
    model.load_state_dict(torch.load(weights_path, map_location=device))
    model = model.to(device)
    model.eval()
    return model

# Function to ensemble predict
def ensemble_predict(models, transforms_dict, image_path, weights=None, show_vis=True):
    """
    Predicts region ID using ensemble of models
    
    Parameters:
    - models: dictionary of {model_name: model}
    - transforms_dict: dictionary of {model_name: transform}
    - image_path: path to the test image
    - weights: optional list of weights for each model
    - show_vis: whether to show visualization
    
    Returns:
    - predicted_region: the 1-indexed region ID (as in the original dataset)
    - confidence: confidence of the prediction
    - all_probs: probabilities from all models
    """
    # Load and preprocess the image for each model
    try:
        image = Image.open(image_path).convert('RGB')
    except Exception as e:
        print(f"Error loading image: {e}")
        return None, 0, {}
    
    # Store probabilities from each model
    model_probs = {}
    
    # Process with each model
    for model_name, model in models.items():
        # Get the right transform
        transform_key = model_name if model_name in transforms_dict else "default"
        transform = transforms_dict[transform_key]
        
        # Preprocess image
        img_tensor = transform(image).unsqueeze(0).to(device)
        
        # Get predictions
        with torch.no_grad():
            outputs = model(img_tensor)
            probs = torch.softmax(outputs, dim=1).cpu().numpy()[0]
            model_probs[model_name] = probs
    
    # Convert to list for ensemble
    all_probs = list(model_probs.values())
    
    # Default equal weights if not provided
    if weights is None:
        weights = [1/len(all_probs)] * len(all_probs)
    
    # Compute weighted average
    weighted_probs = np.zeros_like(all_probs[0])
    for i, probs in enumerate(all_probs):
        weighted_probs += weights[i] * probs
    
    # Get prediction and confidence
    pred_idx = np.argmax(weighted_probs)
    confidence = weighted_probs[pred_idx]
    
    # Predicted region is 1-indexed in original dataset
    predicted_region = pred_idx + 1
    
    # Visualize if requested
    if show_vis:
        # Plot the image
        plt.figure(figsize=(12, 8))
        plt.subplot(1, 2, 1)
        plt.imshow(image)
        plt.title(f"Predicted: Region {predicted_region} (Confidence: {confidence:.2f})")
        plt.axis('off')
        
        # Plot the probabilities
        plt.subplot(1, 2, 2)
        bar_positions = np.arange(len(weighted_probs))
        plt.bar(bar_positions, weighted_probs)
        plt.xlabel('Region ID')
        plt.ylabel('Probability')
        plt.title('Region Probabilities')
        plt.xticks(bar_positions, [f"R{i+1}" for i in range(len(weighted_probs))])
        
        # Highlight the predicted region
        plt.bar(pred_idx, weighted_probs[pred_idx], color='red')
        
        plt.tight_layout()
        plt.savefig('prediction_result.png')
        plt.show()
    
    return predicted_region, confidence, model_probs

# Function to process multiple images in a directory
def batch_predict(models, transforms_dict, image_dir, weights=None, output_file=None):
    """
    Process all images in a directory and save results
    """
    results = {}
    
    # Get all image files
    image_files = [f for f in os.listdir(image_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff'))]
    
    for img_file in image_files:
        img_path = os.path.join(image_dir, img_file)
        print(f"Processing {img_path}...")
        
        region_id, confidence, _ = ensemble_predict(
            models, transforms_dict, img_path, weights, show_vis=False
        )
        
        results[img_file] = {
            "region_id": int(region_id),
            "confidence": float(confidence)
        }
    
    # Save results if requested
    if output_file:
        with open(output_file, 'w') as f:
            json.dump(results, f, indent=2)
        print(f"Results saved to {output_file}")
    
    return results

def main():
    # Parse command line arguments
    parser = argparse.ArgumentParser(description='Predict Region ID from image')
    parser.add_argument('--image', type=str, help='Path to the test image')
    parser.add_argument('--image_dir', type=str, help='Path to directory containing images')
    parser.add_argument('--models_dir', type=str, default='.', help='Directory containing model weights')
    parser.add_argument('--output', type=str, help='Output JSON file for batch predictions')
    parser.add_argument('--no_visualization', action='store_true', help='Disable visualization')
    args = parser.parse_args()
    
    # Model configurations
    model_configs = [
        {"name": "convnext", "weights": f"{args.models_dir}/convnext_region_classifier.pth", "transform_key": "convnext"},
        {"name": "vit", "weights": f"{args.models_dir}/vit_region_classifier.pth", "transform_key": "vit"},
        {"name": "swin", "weights": f"{args.models_dir}/swin_region_classifier.pth", "transform_key": "swin"},
        {"name": "efficientnet", "weights": f"{args.models_dir}/efficientnet_region_classifier.pth", "transform_key": "default"}
    ]
    
    # Load models
    models = {}
    for config in model_configs:
        try:
            print(f"Loading {config['name']} model from {config['weights']}...")
            models[config['name']] = load_model(config['name'], config['weights'])
        except Exception as e:
            print(f"Error loading {config['name']} model: {e}")
    
    # Define model weights (equal by default, but you can adjust based on validation results)
    # Replace these with your actual validation accuracies from training
    weights = {
        "convnext": 0.26,
        "vit": 0.24,
        "swin": 0.25,
        "efficientnet": 0.25
    }
    
    # Normalize weights to sum to 1 (and adjust for any missing models)
    available_models = list(models.keys())
    available_weights = [weights[name] for name in available_models]
    if sum(available_weights) > 0:
        normalized_weights = [w/sum(available_weights) for w in available_weights]
    else:
        normalized_weights = [1/len(available_models)] * len(available_models)
    
    print(f"Using models: {available_models}")
    print(f"With weights: {normalized_weights}")
    
    # Handle single image or batch prediction
    if args.image:
        # Predict single image
        region_id, confidence, model_probs = ensemble_predict(
            models, val_transforms, args.image, 
            weights=normalized_weights, 
            show_vis=not args.no_visualization
        )
        
        print(f"\nPrediction for {args.image}:")
        print(f"Region ID: {region_id}")
        print(f"Confidence: {confidence:.4f}")
        
        # Show individual model predictions
        print("\nIndividual model predictions:")
        for model_name, probs in model_probs.items():
            pred = np.argmax(probs) + 1
            conf = probs[pred-1]
            print(f"  {model_name}: Region {pred} (Confidence: {conf:.4f})")
    
    elif args.image_dir:
        # Batch predict all images in directory
        results = batch_predict(
            models, val_transforms, args.image_dir,
            weights=normalized_weights,
            output_file=args.output
        )
        
        # Print summary
        print("\nPrediction Summary:")
        for img_name, result in results.items():
            print(f"{img_name}: Region {result['region_id']} (Confidence: {result['confidence']:.4f})")
    
    else:
        print("Please provide either --image or --image_dir argument")

if __name__ == "__main__":
    main()
```

```python
import os
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import torch.nn.functional as F

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Define paths
TRAIN_IMG_DIR = '/kaggle/input/iiith-campus-images-dataset/images_train/images_train/images_train'
VALID_IMG_DIR = '/kaggle/input/iiith-campus-images-dataset/images_val/images_val'
TRAIN_LABELS = '/kaggle/input/iiith-campus-images-dataset/labels_train_updated.csv'
VALID_LABELS = '/kaggle/input/iiith-campus-images-dataset/labels_val_updated.csv'

# Define transformations with more robust preprocessing
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Standard size for most modern CNNs
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Enhanced Dataset class with geographic features
class EnhancedCampusDataset(Dataset):
    def __init__(self, img_dir, labels_file, transform=None, use_geo_features=True):
        self.img_dir = img_dir
        self.transform = transform
        self.use_geo_features = use_geo_features
        self.labels_df = pd.read_csv(labels_file)
        self.valid_indices = []

        # Normalize geographic features if using them
        if use_geo_features:
            for col in ['latitude', 'longitude', 'angle']:
                if col in self.labels_df.columns:
                    # Convert to float and normalize
                    self.labels_df[col] = self.labels_df[col].astype(float)
                    min_val = self.labels_df[col].min()
                    max_val = self.labels_df[col].max()
                    if max_val > min_val:
                        self.labels_df[col] = (self.labels_df[col] - min_val) / (max_val - min_val)

        # Check which images can be loaded successfully
        print("Verifying dataset integrity...")
        for idx in tqdm(range(len(self.labels_df))):
            try:
                img_path = os.path.join(img_dir, self.labels_df.iloc[idx]['filename'])
                if os.path.exists(img_path):
                    Image.open(img_path).convert('RGB')
                    self.valid_indices.append(idx)
            except Exception as e:
                print(f"Skipping image at index {idx}: {e}")

        print(f"Found {len(self.valid_indices)} valid images out of {len(self.labels_df)}")

    def __len__(self):
        return len(self.valid_indices)

    def __getitem__(self, idx):
        true_idx = self.valid_indices[idx]
        img_name = os.path.join(self.img_dir, self.labels_df.iloc[true_idx]['filename'])
        image = Image.open(img_name).convert('RGB')

        region_id = self.labels_df.iloc[true_idx]['Region_ID']

        if self.transform:
            image = self.transform(image)

        # If using geographic features, append them to the output
        if self.use_geo_features and 'latitude' in self.labels_df.columns:
            geo_features = torch.tensor([
                self.labels_df.iloc[true_idx]['latitude'],
                self.labels_df.iloc[true_idx]['longitude'],
                self.labels_df.iloc[true_idx]['angle']
            ], dtype=torch.float32)
            return image, geo_features, region_id - 1
        else:
            return image, region_id - 1

# Load the datasets
try:
    # We'll initially try with geo features
    use_geo_features = True
    try:
        train_dataset = EnhancedCampusDataset(TRAIN_IMG_DIR, TRAIN_LABELS, transform=transform, use_geo_features=use_geo_features)
        valid_dataset = EnhancedCampusDataset(VALID_IMG_DIR, VALID_LABELS, transform=transform, use_geo_features=use_geo_features)
    except KeyError:
        # If geo features columns aren't found, fallback to image-only
        print("Geographic features not found in dataset, using image features only.")
        use_geo_features = False
        train_dataset = EnhancedCampusDataset(TRAIN_IMG_DIR, TRAIN_LABELS, transform=transform, use_geo_features=False)
        valid_dataset = EnhancedCampusDataset(VALID_IMG_DIR, VALID_LABELS, transform=transform, use_geo_features=False)

    print(f"Loaded {len(train_dataset)} training samples and {len(valid_dataset)} validation samples")

    # Create data loaders with a smaller batch size for better generalization
    train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True, num_workers=2)
    valid_loader = DataLoader(valid_dataset, batch_size=16, shuffle=False, num_workers=2)

except Exception as e:
    print(f"Error loading datasets: {e}")
    print("Please check the paths and structure of your data files")

# Define an enhanced model - Using EfficientNet B0 which is more accurate than ResNet18
class EnhancedRegionClassifier(nn.Module):
    def __init__(self, num_classes=15, use_geo_features=True):
        super(EnhancedRegionClassifier, self).__init__()
        # EfficientNet base model (can be swapped with other models)
        self.base_model = models.efficientnet_b0(weights='IMAGENET1K_V1')
        self.use_geo_features = use_geo_features

        # Get the number of features in the last layer
        in_features = self.base_model.classifier[1].in_features
        self.base_model.classifier = nn.Identity()

        # Additional layers
        feature_dim = in_features + 3 if use_geo_features else in_features

        # Create a more complex classifier for better learning
        self.classifier = nn.Sequential(
            nn.Dropout(0.3),
            nn.Linear(feature_dim, 512),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Linear(256, num_classes)
        )

        # Temperature parameter for softmax (used for soft classification)
        self.temperature = nn.Parameter(torch.ones(1) * 1.0)

    def forward(self, x, geo_features=None):
        # Extract features from the image
        features = self.base_model(x)

        # Concatenate with geographic features if available
        if self.use_geo_features and geo_features is not None:
            features = torch.cat((features, geo_features), dim=1)

        # Get logits from the classifier
        logits = self.classifier(features)

        # Apply temperature scaling for softer probabilities
        if self.training:
            return logits
        else:
            return logits / self.temperature

# Create a model instance
model = EnhancedRegionClassifier(num_classes=15, use_geo_features=use_geo_features)
model = model.to(device)

# Freeze early layers for transfer learning
for name, param in model.base_model.named_parameters():
    if "features.0" in name or "features.1" in name:
        param.requires_grad = False

# Use a more sophisticated loss function
criterion = nn.CrossEntropyLoss(label_smoothing=0.1)  # Label smoothing helps with classification

# Use a learning rate scheduler and a better optimizer
optimizer = optim.AdamW(model.parameters(), lr=0.001, weight_decay=1e-4)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2)

# Training function with soft classification
def train_model(model, train_loader, valid_loader, criterion, optimizer, scheduler, num_epochs=15):
    history = {
        'train_loss': [],
        'valid_loss': [],
        'valid_acc': [],
        'valid_top3_acc': []  # For tracking top-3 accuracy
    }

    best_acc = 0.0

    for epoch in range(num_epochs):
        # Training phase
        model.train()
        running_loss = 0.0
        total_samples = 0

        progress_bar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{num_epochs}')
        for batch in progress_bar:
            # Handle both dataset types (with or without geo features)
            if use_geo_features:
                inputs, geo_features, labels = batch
                geo_features = geo_features.to(device)
            else:
                inputs, labels = batch
                geo_features = None

            # Skip empty batches
            if inputs.size(0) == 0:
                continue

            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()

            if use_geo_features:
                outputs = model(inputs, geo_features)
            else:
                outputs = model(inputs)

            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)
            total_samples += inputs.size(0)
            progress_bar.set_postfix({'loss': loss.item()})

        epoch_loss = running_loss / total_samples if total_samples > 0 else 0
        history['train_loss'].append(epoch_loss)

        # Validation phase
        model.eval()
        running_loss = 0.0
        all_preds = []
        all_labels = []
        all_soft_preds = []  # For storing soft predictions
        total_val_samples = 0
        correct_top1 = 0
        correct_top3 = 0

        with torch.no_grad():
            for batch in tqdm(valid_loader, desc='Validation'):
                if use_geo_features:
                    inputs, geo_features, labels = batch
                    geo_features = geo_features.to(device)
                else:
                    inputs, labels = batch
                    geo_features = None

                # Skip empty batches
                if inputs.size(0) == 0:
                    continue

                inputs, labels = inputs.to(device), labels.to(device)

                if use_geo_features:
                    outputs = model(inputs, geo_features)
                else:
                    outputs = model(inputs)

                loss = criterion(outputs, labels)

                running_loss += loss.item() * inputs.size(0)
                total_val_samples += inputs.size(0)

                # Get top-1 predictions
                _, preds = torch.max(outputs, 1)
                all_preds.extend(preds.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())

                # Calculate soft predictions (probabilities)
                soft_preds = F.softmax(outputs, dim=1)
                all_soft_preds.append(soft_preds.cpu())

                # Calculate top-1 and top-3 accuracy
                correct_top1 += (preds == labels).sum().item()

                # Top-3 accuracy
                _, top3_indices = outputs.topk(3, dim=1)
                batch_correct = 0
                for i in range(labels.size(0)):
                    if labels[i] in top3_indices[i]:
                        batch_correct += 1

                correct_top3 += batch_correct

        # Calculate validation metrics
        valid_loss = running_loss / total_val_samples if total_val_samples > 0 else 0
        valid_acc = correct_top1 / total_val_samples if total_val_samples > 0 else 0
        valid_top3_acc = correct_top3 / total_val_samples if total_val_samples > 0 else 0

        history['valid_loss'].append(valid_loss)
        history['valid_acc'].append(valid_acc)
        history['valid_top3_acc'].append(valid_top3_acc)

        # Update learning rate based on validation loss
        scheduler.step(valid_loss)

        print(f'Epoch {epoch+1}/{num_epochs} | '
              f'Train Loss: {epoch_loss:.4f} | '
              f'Valid Loss: {valid_loss:.4f} | '
              f'Valid Acc: {valid_acc:.4f} | '
              f'Valid Top-3 Acc: {valid_top3_acc:.4f}')

        # Save the best model
        if valid_acc > best_acc:
            best_acc = valid_acc
            torch.save(model.state_dict(), 'best_region_classifier_b0_40.pth')
            print(f"Saved new best model with accuracy: {best_acc:.4f}")

    # Combine all soft predictions
    all_soft_preds = torch.cat(all_soft_preds, dim=0).numpy()

    return history, all_preds, all_labels, all_soft_preds

# Train the model with more epochs for better learning
history, final_preds, final_labels, soft_predictions = train_model(
    model=model,
    train_loader=train_loader,
    valid_loader=valid_loader,
    criterion=criterion,
    optimizer=optimizer,
    scheduler=scheduler,
    num_epochs=40
)

# Print final validation accuracy and classification report
print("\nFinal Validation Accuracy:", accuracy_score(final_labels, final_preds))
print("\nClassification Report:")
print(classification_report(final_labels, final_preds,
                          target_names=[f"Region {i+1}" for i in range(15)]))

# Plot training history with top-3 accuracy
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.plot(history['train_loss'], label='Train Loss')
plt.plot(history['valid_loss'], label='Valid Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(history['valid_acc'], label='Top-1 Accuracy')
plt.plot(history['valid_top3_acc'], label='Top-3 Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot learning rate if available
current_lr = optimizer.param_groups[0]['lr']
plt.subplot(1, 3, 3)
plt.bar(0, current_lr)
plt.ylabel('Final Learning Rate')
plt.xticks([])
plt.tight_layout()
plt.show()

# Save the final model
torch.save(model.state_dict(), 'efficientnet_region_classifier_b0_40.pth')
print("Final model saved as 'efficientnet_region_classifier_b0_40.pth'")

# Create confusion matrix
cm = confusion_matrix(final_labels, final_preds)
plt.figure(figsize=(10, 8))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = np.arange(15)
plt.xticks(tick_marks, [f"Region {i+1}" for i in range(15)], rotation=45)
plt.yticks(tick_marks, [f"Region {i+1}" for i in range(15)])
plt.tight_layout()
plt.xlabel('Predicted Region')
plt.ylabel('True Region')
plt.show()

# Analyze soft predictions for boundary cases
def analyze_boundary_cases(soft_preds, true_labels, threshold=0.3):
    """Identify boundary cases where the model is uncertain"""
    boundary_cases = []
    top2_diffs = []

    for i, probs in enumerate(soft_preds):
        sorted_probs, indices = torch.from_numpy(probs).sort(descending=True)
        # If difference between top two probabilities is small, it's a boundary case
        if sorted_probs[0] - sorted_probs[1] < threshold:
            boundary_cases.append(i)
            top2_diffs.append((sorted_probs[0] - sorted_probs[1]).item())

    print(f"\nFound {len(boundary_cases)} boundary cases out of {len(soft_preds)} ({len(boundary_cases)/len(soft_preds)*100:.2f}%)")

    if len(boundary_cases) > 0:
        plt.figure(figsize=(10, 6))
        plt.hist(top2_diffs, bins=20)
        plt.title('Distribution of Top-2 Probability Differences for Boundary Cases')
        plt.xlabel('Difference between top two probabilities')
        plt.ylabel('Count')
        plt.tight_layout()
        plt.savefig('boundary_cases_distribution.png')
        plt.show()

    return boundary_cases

# Analyze boundary cases
boundary_indices = analyze_boundary_cases(soft_predictions, final_labels)

```

```python
import os
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import torch.nn.functional as F

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Define paths
TRAIN_IMG_DIR = '/kaggle/input/iiith-campus-images-dataset/images_train'
VALID_IMG_DIR = '/kaggle/input/iiith-campus-images-dataset/images_val'
TRAIN_LABELS = '/kaggle/input/iiith-campus-images-dataset/labels_train_updated.csv'
VALID_LABELS = '/kaggle/input/iiith-campus-images-dataset/labels_val_updated.csv'

# Define transformations with more robust preprocessing
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Standard size for most modern CNNs
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Enhanced Dataset class with geographic features
class EnhancedCampusDataset(Dataset):
    def __init__(self, img_dir, labels_file, transform=None, use_geo_features=True):
        self.img_dir = img_dir
        self.transform = transform
        self.use_geo_features = use_geo_features
        self.labels_df = pd.read_csv(labels_file)
        self.valid_indices = []

        # Normalize geographic features if using them
        if use_geo_features:
            for col in ['latitude', 'longitude', 'angle']:
                if col in self.labels_df.columns:
                    # Convert to float and normalize
                    self.labels_df[col] = self.labels_df[col].astype(float)
                    min_val = self.labels_df[col].min()
                    max_val = self.labels_df[col].max()
                    if max_val > min_val:
                        self.labels_df[col] = (self.labels_df[col] - min_val) / (max_val - min_val)

        # Check which images can be loaded successfully
        print("Verifying dataset integrity...")
        for idx in tqdm(range(len(self.labels_df))):
            try:
                img_path = os.path.join(img_dir, self.labels_df.iloc[idx]['filename'])
                if os.path.exists(img_path):
                    Image.open(img_path).convert('RGB')
                    self.valid_indices.append(idx)
            except Exception as e:
                print(f"Skipping image at index {idx}: {e}")

        print(f"Found {len(self.valid_indices)} valid images out of {len(self.labels_df)}")

    def __len__(self):
        return len(self.valid_indices)

    def __getitem__(self, idx):
        true_idx = self.valid_indices[idx]
        img_name = os.path.join(self.img_dir, self.labels_df.iloc[true_idx]['filename'])
        image = Image.open(img_name).convert('RGB')

        region_id = self.labels_df.iloc[true_idx]['Region_ID']

        if self.transform:
            image = self.transform(image)

        # If using geographic features, append them to the output
        if self.use_geo_features and 'latitude' in self.labels_df.columns:
            geo_features = torch.tensor([
                self.labels_df.iloc[true_idx]['latitude'],
                self.labels_df.iloc[true_idx]['longitude'],
                self.labels_df.iloc[true_idx]['angle']
            ], dtype=torch.float32)
            return image, geo_features, region_id - 1
        else:
            return image, region_id - 1

# Load the datasets
try:
    # We'll initially try with geo features
    use_geo_features = True
    try:
        train_dataset = EnhancedCampusDataset(TRAIN_IMG_DIR, TRAIN_LABELS, transform=transform, use_geo_features=use_geo_features)
        valid_dataset = EnhancedCampusDataset(VALID_IMG_DIR, VALID_LABELS, transform=transform, use_geo_features=use_geo_features)
    except KeyError:
        # If geo features columns aren't found, fallback to image-only
        print("Geographic features not found in dataset, using image features only.")
        use_geo_features = False
        train_dataset = EnhancedCampusDataset(TRAIN_IMG_DIR, TRAIN_LABELS, transform=transform, use_geo_features=False)
        valid_dataset = EnhancedCampusDataset(VALID_IMG_DIR, VALID_LABELS, transform=transform, use_geo_features=False)

    print(f"Loaded {len(train_dataset)} training samples and {len(valid_dataset)} validation samples")

    # Create data loaders with a smaller batch size for better generalization
    train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True, num_workers=2)
    valid_loader = DataLoader(valid_dataset, batch_size=16, shuffle=False, num_workers=2)

except Exception as e:
    print(f"Error loading datasets: {e}")
    print("Please check the paths and structure of your data files")

# Define an enhanced model - Using EfficientNet B0 which is more accurate than ResNet18
class EnhancedRegionClassifier(nn.Module):
    def __init__(self, num_classes=15, use_geo_features=True):
        super(EnhancedRegionClassifier, self).__init__()
        # EfficientNet base model (can be swapped with other models)
        self.base_model = models.efficientnet_b1(weights='IMAGENET1K_V1')
        self.use_geo_features = use_geo_features

        # Get the number of features in the last layer
        in_features = self.base_model.classifier[1].in_features
        self.base_model.classifier = nn.Identity()

        # Additional layers
        feature_dim = in_features + 3 if use_geo_features else in_features

        # Create a more complex classifier for better learning
        self.classifier = nn.Sequential(
            nn.Dropout(0.3),
            nn.Linear(feature_dim, 512),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Linear(256, num_classes)
        )

        # Temperature parameter for softmax (used for soft classification)
        self.temperature = nn.Parameter(torch.ones(1) * 1.0)

    def forward(self, x, geo_features=None):
        # Extract features from the image
        features = self.base_model(x)

        # Concatenate with geographic features if available
        if self.use_geo_features and geo_features is not None:
            features = torch.cat((features, geo_features), dim=1)

        # Get logits from the classifier
        logits = self.classifier(features)

        # Apply temperature scaling for softer probabilities
        if self.training:
            return logits
        else:
            return logits / self.temperature

# Create a model instance
model = EnhancedRegionClassifier(num_classes=15, use_geo_features=use_geo_features)
model = model.to(device)

# Freeze early layers for transfer learning
for name, param in model.base_model.named_parameters():
    if "features.0" in name or "features.1" in name:
        param.requires_grad = False

# Use a more sophisticated loss function
criterion = nn.CrossEntropyLoss(label_smoothing=0.1)  # Label smoothing helps with classification

# Use a learning rate scheduler and a better optimizer
optimizer = optim.AdamW(model.parameters(), lr=0.001, weight_decay=1e-4)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2)

# Training function with soft classification
def train_model(model, train_loader, valid_loader, criterion, optimizer, scheduler, num_epochs=15):
    history = {
        'train_loss': [],
        'valid_loss': [],
        'valid_acc': [],
        'valid_top3_acc': []  # For tracking top-3 accuracy
    }

    best_acc = 0.0

    for epoch in range(num_epochs):
        # Training phase
        model.train()
        running_loss = 0.0
        total_samples = 0

        progress_bar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{num_epochs}')
        for batch in progress_bar:
            # Handle both dataset types (with or without geo features)
            if use_geo_features:
                inputs, geo_features, labels = batch
                geo_features = geo_features.to(device)
            else:
                inputs, labels = batch
                geo_features = None

            # Skip empty batches
            if inputs.size(0) == 0:
                continue

            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()

            if use_geo_features:
                outputs = model(inputs, geo_features)
            else:
                outputs = model(inputs)

            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)
            total_samples += inputs.size(0)
            progress_bar.set_postfix({'loss': loss.item()})

        epoch_loss = running_loss / total_samples if total_samples > 0 else 0
        history['train_loss'].append(epoch_loss)

        # Validation phase
        model.eval()
        running_loss = 0.0
        all_preds = []
        all_labels = []
        all_soft_preds = []  # For storing soft predictions
        total_val_samples = 0
        correct_top1 = 0
        correct_top3 = 0

        with torch.no_grad():
            for batch in tqdm(valid_loader, desc='Validation'):
                if use_geo_features:
                    inputs, geo_features, labels = batch
                    geo_features = geo_features.to(device)
                else:
                    inputs, labels = batch
                    geo_features = None

                # Skip empty batches
                if inputs.size(0) == 0:
                    continue

                inputs, labels = inputs.to(device), labels.to(device)

                if use_geo_features:
                    outputs = model(inputs, geo_features)
                else:
                    outputs = model(inputs)

                loss = criterion(outputs, labels)

                running_loss += loss.item() * inputs.size(0)
                total_val_samples += inputs.size(0)

                # Get top-1 predictions
                _, preds = torch.max(outputs, 1)
                all_preds.extend(preds.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())

                # Calculate soft predictions (probabilities)
                soft_preds = F.softmax(outputs, dim=1)
                all_soft_preds.append(soft_preds.cpu())

                # Calculate top-1 and top-3 accuracy
                correct_top1 += (preds == labels).sum().item()

                # Top-3 accuracy
                _, top3_indices = outputs.topk(3, dim=1)
                batch_correct = 0
                for i in range(labels.size(0)):
                    if labels[i] in top3_indices[i]:
                        batch_correct += 1

                correct_top3 += batch_correct

        # Calculate validation metrics
        valid_loss = running_loss / total_val_samples if total_val_samples > 0 else 0
        valid_acc = correct_top1 / total_val_samples if total_val_samples > 0 else 0
        valid_top3_acc = correct_top3 / total_val_samples if total_val_samples > 0 else 0

        history['valid_loss'].append(valid_loss)
        history['valid_acc'].append(valid_acc)
        history['valid_top3_acc'].append(valid_top3_acc)

        # Update learning rate based on validation loss
        scheduler.step(valid_loss)

        print(f'Epoch {epoch+1}/{num_epochs} | '
              f'Train Loss: {epoch_loss:.4f} | '
              f'Valid Loss: {valid_loss:.4f} | '
              f'Valid Acc: {valid_acc:.4f} | '
              f'Valid Top-3 Acc: {valid_top3_acc:.4f}')

        # Save the best model
        if valid_acc > best_acc:
            best_acc = valid_acc
            torch.save(model.state_dict(), 'best_region_classifier_b1_40.pth')
            print(f"Saved new best model with accuracy: {best_acc:.4f}")

    # Combine all soft predictions
    all_soft_preds = torch.cat(all_soft_preds, dim=0).numpy()

    return history, all_preds, all_labels, all_soft_preds

# Train the model with more epochs for better learning
history, final_preds, final_labels, soft_predictions = train_model(
    model=model,
    train_loader=train_loader,
    valid_loader=valid_loader,
    criterion=criterion,
    optimizer=optimizer,
    scheduler=scheduler,
    num_epochs=40
)

# Print final validation accuracy and classification report
print("\nFinal Validation Accuracy:", accuracy_score(final_labels, final_preds))
print("\nClassification Report:")
print(classification_report(final_labels, final_preds,
                          target_names=[f"Region {i+1}" for i in range(15)]))

# Plot training history with top-3 accuracy
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.plot(history['train_loss'], label='Train Loss')
plt.plot(history['valid_loss'], label='Valid Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(history['valid_acc'], label='Top-1 Accuracy')
plt.plot(history['valid_top3_acc'], label='Top-3 Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot learning rate if available
current_lr = optimizer.param_groups[0]['lr']
plt.subplot(1, 3, 3)
plt.bar(0, current_lr)
plt.ylabel('Final Learning Rate')
plt.xticks([])
plt.tight_layout()
plt.show()

# Save the final model
torch.save(model.state_dict(), 'efficientnet_region_classifier_b1_40.pth')
print("Final model saved as 'efficientnet_region_classifier_b1_40.pth'")

# Create confusion matrix
cm = confusion_matrix(final_labels, final_preds)
plt.figure(figsize=(10, 8))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = np.arange(15)
plt.xticks(tick_marks, [f"Region {i+1}" for i in range(15)], rotation=45)
plt.yticks(tick_marks, [f"Region {i+1}" for i in range(15)])
plt.tight_layout()
plt.xlabel('Predicted Region')
plt.ylabel('True Region')
plt.show()

# Analyze soft predictions for boundary cases
def analyze_boundary_cases(soft_preds, true_labels, threshold=0.3):
    """Identify boundary cases where the model is uncertain"""
    boundary_cases = []
    top2_diffs = []

    for i, probs in enumerate(soft_preds):
        sorted_probs, indices = torch.from_numpy(probs).sort(descending=True)
        # If difference between top two probabilities is small, it's a boundary case
        if sorted_probs[0] - sorted_probs[1] < threshold:
            boundary_cases.append(i)
            top2_diffs.append((sorted_probs[0] - sorted_probs[1]).item())

    print(f"\nFound {len(boundary_cases)} boundary cases out of {len(soft_preds)} ({len(boundary_cases)/len(soft_preds)*100:.2f}%)")

    if len(boundary_cases) > 0:
        plt.figure(figsize=(10, 6))
        plt.hist(top2_diffs, bins=20)
        plt.title('Distribution of Top-2 Probability Differences for Boundary Cases')
        plt.xlabel('Difference between top two probabilities')
        plt.ylabel('Count')
        plt.tight_layout()
        plt.show()

    return boundary_cases

# Analyze boundary cases
boundary_indices = analyze_boundary_cases(soft_predictions, final_labels)

```

```python
import os
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import torch.nn.functional as F

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Define paths
TRAIN_IMG_DIR = '/kaggle/input/iiith-campus-images-dataset/images_train'
VALID_IMG_DIR = '/kaggle/input/iiith-campus-images-dataset/images_val'
TRAIN_LABELS = '/kaggle/input/iiith-campus-images-dataset/labels_train_updated.csv'
VALID_LABELS = '/kaggle/input/iiith-campus-images-dataset/labels_val_updated.csv'

# Define transformations with more robust preprocessing
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Standard size for most modern CNNs
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Enhanced Dataset class with geographic features
class EnhancedCampusDataset(Dataset):
    def __init__(self, img_dir, labels_file, transform=None, use_geo_features=True):
        self.img_dir = img_dir
        self.transform = transform
        self.use_geo_features = use_geo_features
        self.labels_df = pd.read_csv(labels_file)
        self.valid_indices = []

        # Normalize geographic features if using them
        if use_geo_features:
            for col in ['latitude', 'longitude', 'angle']:
                if col in self.labels_df.columns:
                    # Convert to float and normalize
                    self.labels_df[col] = self.labels_df[col].astype(float)
                    min_val = self.labels_df[col].min()
                    max_val = self.labels_df[col].max()
                    if max_val > min_val:
                        self.labels_df[col] = (self.labels_df[col] - min_val) / (max_val - min_val)

        # Check which images can be loaded successfully
        print("Verifying dataset integrity...")
        for idx in tqdm(range(len(self.labels_df))):
            try:
                img_path = os.path.join(img_dir, self.labels_df.iloc[idx]['filename'])
                if os.path.exists(img_path):
                    Image.open(img_path).convert('RGB')
                    self.valid_indices.append(idx)
            except Exception as e:
                print(f"Skipping image at index {idx}: {e}")

        print(f"Found {len(self.valid_indices)} valid images out of {len(self.labels_df)}")

    def __len__(self):
        return len(self.valid_indices)

    def __getitem__(self, idx):
        true_idx = self.valid_indices[idx]
        img_name = os.path.join(self.img_dir, self.labels_df.iloc[true_idx]['filename'])
        image = Image.open(img_name).convert('RGB')

        region_id = self.labels_df.iloc[true_idx]['Region_ID']

        if self.transform:
            image = self.transform(image)

        # If using geographic features, append them to the output
        if self.use_geo_features and 'latitude' in self.labels_df.columns:
            geo_features = torch.tensor([
                self.labels_df.iloc[true_idx]['latitude'],
                self.labels_df.iloc[true_idx]['longitude'],
                self.labels_df.iloc[true_idx]['angle']
            ], dtype=torch.float32)
            return image, geo_features, region_id - 1
        else:
            return image, region_id - 1

# Load the datasets
try:
    # We'll initially try with geo features
    use_geo_features = True
    try:
        train_dataset = EnhancedCampusDataset(TRAIN_IMG_DIR, TRAIN_LABELS, transform=transform, use_geo_features=use_geo_features)
        valid_dataset = EnhancedCampusDataset(VALID_IMG_DIR, VALID_LABELS, transform=transform, use_geo_features=use_geo_features)
    except KeyError:
        # If geo features columns aren't found, fallback to image-only
        print("Geographic features not found in dataset, using image features only.")
        use_geo_features = False
        train_dataset = EnhancedCampusDataset(TRAIN_IMG_DIR, TRAIN_LABELS, transform=transform, use_geo_features=False)
        valid_dataset = EnhancedCampusDataset(VALID_IMG_DIR, VALID_LABELS, transform=transform, use_geo_features=False)

    print(f"Loaded {len(train_dataset)} training samples and {len(valid_dataset)} validation samples")

    # Create data loaders with a smaller batch size for better generalization
    train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True, num_workers=2)
    valid_loader = DataLoader(valid_dataset, batch_size=16, shuffle=False, num_workers=2)

except Exception as e:
    print(f"Error loading datasets: {e}")
    print("Please check the paths and structure of your data files")

# Define an enhanced model - Using EfficientNet B0 which is more accurate than ResNet18
class EnhancedRegionClassifier(nn.Module):
    def __init__(self, num_classes=15, use_geo_features=True):
        super(EnhancedRegionClassifier, self).__init__()
        # EfficientNet base model (can be swapped with other models)
        self.base_model = models.efficientnet_b2(weights='IMAGENET1K_V1')
        self.use_geo_features = use_geo_features

        # Get the number of features in the last layer
        in_features = self.base_model.classifier[1].in_features
        self.base_model.classifier = nn.Identity()

        # Additional layers
        feature_dim = in_features + 3 if use_geo_features else in_features

        # Create a more complex classifier for better learning
        self.classifier = nn.Sequential(
            nn.Dropout(0.3),
            nn.Linear(feature_dim, 512),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(512, 256),
            nn.ReLU(),
            nn.Linear(256, num_classes)
        )

        # Temperature parameter for softmax (used for soft classification)
        self.temperature = nn.Parameter(torch.ones(1) * 1.0)

    def forward(self, x, geo_features=None):
        # Extract features from the image
        features = self.base_model(x)

        # Concatenate with geographic features if available
        if self.use_geo_features and geo_features is not None:
            features = torch.cat((features, geo_features), dim=1)

        # Get logits from the classifier
        logits = self.classifier(features)

        # Apply temperature scaling for softer probabilities
        if self.training:
            return logits
        else:
            return logits / self.temperature

# Create a model instance
model = EnhancedRegionClassifier(num_classes=15, use_geo_features=use_geo_features)
model = model.to(device)

# Freeze early layers for transfer learning
for name, param in model.base_model.named_parameters():
    if "features.0" in name or "features.1" in name:
        param.requires_grad = False

# Use a more sophisticated loss function
criterion = nn.CrossEntropyLoss(label_smoothing=0.1)  # Label smoothing helps with classification

# Use a learning rate scheduler and a better optimizer
optimizer = optim.AdamW(model.parameters(), lr=0.001, weight_decay=1e-4)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=2)

# Training function with soft classification
def train_model(model, train_loader, valid_loader, criterion, optimizer, scheduler, num_epochs=15):
    history = {
        'train_loss': [],
        'valid_loss': [],
        'valid_acc': [],
        'valid_top3_acc': []  # For tracking top-3 accuracy
    }

    best_acc = 0.0

    for epoch in range(num_epochs):
        # Training phase
        model.train()
        running_loss = 0.0
        total_samples = 0

        progress_bar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{num_epochs}')
        for batch in progress_bar:
            # Handle both dataset types (with or without geo features)
            if use_geo_features:
                inputs, geo_features, labels = batch
                geo_features = geo_features.to(device)
            else:
                inputs, labels = batch
                geo_features = None

            # Skip empty batches
            if inputs.size(0) == 0:
                continue

            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()

            if use_geo_features:
                outputs = model(inputs, geo_features)
            else:
                outputs = model(inputs)

            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * inputs.size(0)
            total_samples += inputs.size(0)
            progress_bar.set_postfix({'loss': loss.item()})

        epoch_loss = running_loss / total_samples if total_samples > 0 else 0
        history['train_loss'].append(epoch_loss)

        # Validation phase
        model.eval()
        running_loss = 0.0
        all_preds = []
        all_labels = []
        all_soft_preds = []  # For storing soft predictions
        total_val_samples = 0
        correct_top1 = 0
        correct_top3 = 0

        with torch.no_grad():
            for batch in tqdm(valid_loader, desc='Validation'):
                if use_geo_features:
                    inputs, geo_features, labels = batch
                    geo_features = geo_features.to(device)
                else:
                    inputs, labels = batch
                    geo_features = None

                # Skip empty batches
                if inputs.size(0) == 0:
                    continue

                inputs, labels = inputs.to(device), labels.to(device)

                if use_geo_features:
                    outputs = model(inputs, geo_features)
                else:
                    outputs = model(inputs)

                loss = criterion(outputs, labels)

                running_loss += loss.item() * inputs.size(0)
                total_val_samples += inputs.size(0)

                # Get top-1 predictions
                _, preds = torch.max(outputs, 1)
                all_preds.extend(preds.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())

                # Calculate soft predictions (probabilities)
                soft_preds = F.softmax(outputs, dim=1)
                all_soft_preds.append(soft_preds.cpu())

                # Calculate top-1 and top-3 accuracy
                correct_top1 += (preds == labels).sum().item()

                # Top-3 accuracy
                _, top3_indices = outputs.topk(3, dim=1)
                batch_correct = 0
                for i in range(labels.size(0)):
                    if labels[i] in top3_indices[i]:
                        batch_correct += 1

                correct_top3 += batch_correct

        # Calculate validation metrics
        valid_loss = running_loss / total_val_samples if total_val_samples > 0 else 0
        valid_acc = correct_top1 / total_val_samples if total_val_samples > 0 else 0
        valid_top3_acc = correct_top3 / total_val_samples if total_val_samples > 0 else 0

        history['valid_loss'].append(valid_loss)
        history['valid_acc'].append(valid_acc)
        history['valid_top3_acc'].append(valid_top3_acc)

        # Update learning rate based on validation loss
        scheduler.step(valid_loss)

        print(f'Epoch {epoch+1}/{num_epochs} | '
              f'Train Loss: {epoch_loss:.4f} | '
              f'Valid Loss: {valid_loss:.4f} | '
              f'Valid Acc: {valid_acc:.4f} | '
              f'Valid Top-3 Acc: {valid_top3_acc:.4f}')

        # Save the best model
        if valid_acc > best_acc:
            best_acc = valid_acc
            torch.save(model.state_dict(), 'best_region_classifier_b2_40.pth')
            print(f"Saved new best model with accuracy: {best_acc:.4f}")

    # Combine all soft predictions
    all_soft_preds = torch.cat(all_soft_preds, dim=0).numpy()

    return history, all_preds, all_labels, all_soft_preds

# Train the model with more epochs for better learning
history, final_preds, final_labels, soft_predictions = train_model(
    model=model,
    train_loader=train_loader,
    valid_loader=valid_loader,
    criterion=criterion,
    optimizer=optimizer,
    scheduler=scheduler,
    num_epochs=40
)

# Print final validation accuracy and classification report
print("\nFinal Validation Accuracy:", accuracy_score(final_labels, final_preds))
print("\nClassification Report:")
print(classification_report(final_labels, final_preds,
                          target_names=[f"Region {i+1}" for i in range(15)]))

# Plot training history with top-3 accuracy
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.plot(history['train_loss'], label='Train Loss')
plt.plot(history['valid_loss'], label='Valid Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(history['valid_acc'], label='Top-1 Accuracy')
plt.plot(history['valid_top3_acc'], label='Top-3 Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

# Plot learning rate if available
current_lr = optimizer.param_groups[0]['lr']
plt.subplot(1, 3, 3)
plt.bar(0, current_lr)
plt.ylabel('Final Learning Rate')
plt.xticks([])
plt.tight_layout()
plt.show()

# Save the final model
torch.save(model.state_dict(), 'efficientnet_region_classifier_b2_40.pth')
print("Final model saved as 'efficientnet_region_classifier_b2_40.pth'")

# Create confusion matrix
cm = confusion_matrix(final_labels, final_preds)
plt.figure(figsize=(10, 8))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = np.arange(15)
plt.xticks(tick_marks, [f"Region {i+1}" for i in range(15)], rotation=45)
plt.yticks(tick_marks, [f"Region {i+1}" for i in range(15)])
plt.tight_layout()
plt.xlabel('Predicted Region')
plt.ylabel('True Region')
plt.show()

# Analyze soft predictions for boundary cases
def analyze_boundary_cases(soft_preds, true_labels, threshold=0.3):
    """Identify boundary cases where the model is uncertain"""
    boundary_cases = []
    top2_diffs = []

    for i, probs in enumerate(soft_preds):
        sorted_probs, indices = torch.from_numpy(probs).sort(descending=True)
        # If difference between top two probabilities is small, it's a boundary case
        if sorted_probs[0] - sorted_probs[1] < threshold:
            boundary_cases.append(i)
            top2_diffs.append((sorted_probs[0] - sorted_probs[1]).item())

    print(f"\nFound {len(boundary_cases)} boundary cases out of {len(soft_preds)} ({len(boundary_cases)/len(soft_preds)*100:.2f}%)")

    if len(boundary_cases) > 0:
        plt.figure(figsize=(10, 6))
        plt.hist(top2_diffs, bins=20)
        plt.title('Distribution of Top-2 Probability Differences for Boundary Cases')
        plt.xlabel('Difference between top two probabilities')
        plt.ylabel('Count')
        plt.tight_layout()
        plt.show()

    return boundary_cases

# Analyze boundary cases
boundary_indices = analyze_boundary_cases(soft_predictions, final_labels)

```
